# Realizable (Problem-Space) Attacks on MNIST Classifier
# ===============================================

#### This notebook demonstrates how to implement realizable adversarial attacks
#### against a deployed MNIST classifier, following the concepts described in
#### the Security and Privacy of AI Knowledge Guide (Section 3.5).


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
import requests
import json
import pickle
import os
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
from scipy import ndimage
import cv2

# For reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# API endpoints for the deployed model
API_URL = "http://localhost:5000"
INFO_ENDPOINT = f"{API_URL}/info"
PREDICT_ENDPOINT = f"{API_URL}/predict_raw"
GRADIENT_ENDPOINT = f"{API_URL}/get_gradient"
```


# ===============================================
# 1. Load and prepare the MNIST dataset
# ===============================================


```python

print("Loading MNIST dataset...")
# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Print dataset shapes
print(f"Training data shape: {x_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Test data shape: {x_test.shape}")
print(f"Test labels shape: {y_test.shape}")

# Preprocess the data (normalize pixel values between 0 and 1)
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Reshape to include channel dimension (MNIST is grayscale, so 1 channel)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Convert class vectors to binary class matrices (one-hot encoding)
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print("Data preprocessing completed.")

# Function to visualize sample images
def visualize_samples(images, labels, title, num_samples=25):
    plt.figure(figsize=(10, 10))
    for i in range(min(num_samples, len(images))):
        plt.subplot(5, 5, i+1)
        # Reshape if the image has a channel dimension
        if len(images[i].shape) == 3:
            plt.imshow(images[i].reshape(28, 28), cmap='gray')
        else:
            plt.imshow(images[i], cmap='gray')
        plt.title(f"Label: {labels[i]}")
        plt.axis('off')
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

# Display random samples from the test set
indices = np.random.choice(len(x_test), 25, replace=False)
visualize_samples(x_test[indices], y_test[indices], "Sample Test Images")
```


# ===============================================
# 2. Test Connection to the Deployed Model
# ===============================================


```python

def get_model_info():
    """Get information about the deployed model"""
    try:
        response = requests.get(INFO_ENDPOINT)
        if response.status_code == 200:
            return response.json()
        else:
            print(f"Error: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error connecting to the model API: {e}")
        return None

def predict_digit(image_data):
    """
    Send an image to the deployed model for prediction
    
    Args:
        image_data: Numpy array of the image (28x28x1)
        
    Returns:
        The prediction results
    """
    # Flatten the image data
    pixels = image_data.flatten().tolist()
    
    # Create request payload
    payload = {
        "pixels": pixels
    }
    
    # Send request to the predict endpoint
    try:
        response = requests.post(PREDICT_ENDPOINT, json=payload)
        if response.status_code == 200:
            return response.json()
        else:
            print(f"Error: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error sending prediction request: {e}")
        return None

def get_gradient(image_data, label):
    """
    Get the gradient of the loss with respect to the input
    
    Args:
        image_data: Numpy array of the image (28x28x1)
        label: The true label of the image
        
    Returns:
        The gradient information
    """
    # Flatten the image data
    pixels = image_data.flatten().tolist()
    
    # Create request payload
    payload = {
        "pixels": pixels,
        "label": int(label)
    }
    
    # Send request to the gradient endpoint
    try:
        response = requests.post(GRADIENT_ENDPOINT, json=payload)
        if response.status_code == 200:
            return response.json()
        else:
            print(f"Error: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error getting gradient: {e}")
        return None

# Test connection to the model API
print("\nTesting connection to the deployed model...")
model_info = get_model_info()
if model_info:
    print("Successfully connected to the model API!")
    print(f"Model name: {model_info.get('model_name', 'Unknown')}")
    print(f"Model accuracy: {model_info.get('test_accuracy', 'Unknown')}")
    print(f"Classes: {model_info.get('classes', 'Unknown')}")
else:
    print("Failed to connect to the model API. Please make sure the model is running at", API_URL)

# Test prediction with a sample image
if model_info:
    print("\nTesting prediction with a sample image...")
    sample_idx = np.random.choice(len(x_test))
    sample_image = x_test[sample_idx]
    sample_label = y_test[sample_idx]
    
    # Display the sample image
    plt.figure(figsize=(3, 3))
    plt.imshow(sample_image.reshape(28, 28), cmap='gray')
    plt.title(f"True Label: {sample_label}")
    plt.axis('off')
    plt.show()
    
    # Get prediction from the deployed model
    prediction_result = predict_digit(sample_image)
    if prediction_result:
        predicted_class = prediction_result.get('predicted_class')
        confidence = prediction_result.get('confidence')
        print(f"Predicted Class: {predicted_class}")
        print(f"Confidence: {confidence:.4f}")
    else:
        print("Failed to get prediction from the model API.")
```


# ===============================================
# 3. Understanding Realizable (Problem-Space) Attacks
# ===============================================


## What are Realizable (Problem-Space) Attacks?

As described in the Security and Privacy of AI Knowledge Guide (Section 3.5), 
realizable (or problem-space) attacks focus on creating adversarial examples that
are valid in the real world, considering practical constraints and limitations.

### Key Characteristics:

1. **Problem Space vs. Feature Space**:
   - Feature space: The mathematical representation used by ML models
   - Problem space: The real-world domain where objects exist with physical constraints

2. **Practical Constraints**:
   - Semantic Preservation: Modified examples must maintain their original functionality
   - Plausibility: Modified examples must appear realistic
   - Robustness: Adversarial modifications should be robust to pre-processing

3. **Side-Effect Features**:
   - Modifications introduce additional changes ("side effects") beyond the targeted alterations
   - These side-effect features are necessary to create valid objects in the problem space

### Why Realizable Attacks Matter:

Traditional adversarial attacks often produce examples that aren't valid in the real world.
For instance, pixel-level modifications to images might look valid to human eyes but wouldn't
transfer to physical objects. Similarly, modifications to code or malware that only affect
feature representations might not yield executable programs.

In this notebook, we'll implement realizable attacks on MNIST digits that preserve 
key characteristics of handwritten digits while causing misclassification.


# ===============================================
# 4. Implementing Realizable Transformations for MNIST
# ===============================================


## Realizable Transformations

For MNIST digits, we'll implement transformations that could occur in real handwritten digits:

1. **Rotation**: Simulate different writing angles
2. **Thickening/Thinning**: Simulate different pen thicknesses
3. **Elastic Deformation**: Simulate different writing styles
4. **Image Translation**: Simulate different positioning on paper
5. **Noise Addition**: Simulate imperfections in paper or scanning

These transformations represent changes that are natural in handwritten digits
and preserve the semantic meaning of the digit, making them realizable.



```python

# Define realizable transformations
def rotate_image(image, angle):
    """Rotate image by a given angle"""
    rotated = ndimage.rotate(image.reshape(28, 28), angle, reshape=False)
    return rotated.reshape(1, 28, 28, 1)

def adjust_thickness(image, factor):
    """Adjust the thickness of pen strokes"""
    # Threshold the image to isolate the digit
    img = image.reshape(28, 28)
    threshold = 0.2
    binary = (img > threshold).astype(np.float32)
    
    # Dilate or erode based on factor
    if factor > 1:  # Thicken
        kernel_size = int(factor)
        kernel = np.ones((kernel_size, kernel_size), np.uint8)
        result = cv2.dilate(binary, kernel, iterations=1)
    else:  # Thin
        kernel_size = int(1/factor)
        kernel = np.ones((kernel_size, kernel_size), np.uint8)
        result = cv2.erode(binary, kernel, iterations=1)
    
    return result.reshape(1, 28, 28, 1)

def elastic_deformation(image, alpha=8, sigma=3):
    """Apply elastic deformation to simulate handwriting style variations"""
    img = image.reshape(28, 28)
    
    # Create random displacement fields
    dx = np.random.uniform(-1, 1, (28, 28)) * alpha
    dy = np.random.uniform(-1, 1, (28, 28)) * alpha
    
    # Gaussian filter for smoothing
    dx = ndimage.gaussian_filter(dx, sigma)
    dy = ndimage.gaussian_filter(dy, sigma)
    
    # Create mesh grid
    y, x = np.meshgrid(np.arange(28), np.arange(28), indexing='ij')
    
    # Displacement field
    indices = np.reshape(y+dy, (-1, 1)), np.reshape(x+dx, (-1, 1))
    
    # Apply displacement
    distorted = ndimage.map_coordinates(img, indices, order=1)
    distorted = distorted.reshape(28, 28)
    
    return distorted.reshape(1, 28, 28, 1)

def translate_image(image, dx, dy):
    """Translate image by dx, dy pixels"""
    translated = ndimage.shift(image.reshape(28, 28), [dy, dx], mode='constant', cval=0)
    return translated.reshape(1, 28, 28, 1)

def add_noise(image, intensity=0.1):
    """Add random noise to the image"""
    noise = np.random.normal(0, intensity, image.shape)
    noisy_image = image + noise
    return np.clip(noisy_image, 0, 1)

# Function to visualize transformations
def visualize_transformations(image, transformations, titles):
    """Visualize various transformations of an image"""
    plt.figure(figsize=(15, 3))
    
    # Original image
    plt.subplot(1, len(transformations)+1, 1)
    plt.imshow(image.reshape(28, 28), cmap='gray')
    plt.title("Original")
    plt.axis('off')
    
    # Transformed images
    for i, (transformed, title) in enumerate(zip(transformations, titles)):
        plt.subplot(1, len(transformations)+1, i+2)
        plt.imshow(transformed.reshape(28, 28), cmap='gray')
        plt.title(title)
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Test the transformations
print("\nDemonstrating realizable transformations:")
test_idx = np.random.choice(len(x_test))
test_image = x_test[test_idx]
test_label = y_test[test_idx]

# Apply transformations
rotated = rotate_image(test_image, 15)
thickened = adjust_thickness(test_image, 2)
thinned = adjust_thickness(test_image, 0.5)
deformed = elastic_deformation(test_image)
translated = translate_image(test_image, 3, -2)
noisy = add_noise(test_image, 0.1)

# Visualize transformations
visualize_transformations(
    test_image,
    [rotated, thickened, thinned, deformed, translated, noisy],
    ["Rotated", "Thickened", "Thinned", "Deformed", "Translated", "Noisy"]
)

# Test if these transformations affect model predictions
def test_transformations_on_model(image, label, transformations, titles):
    """Test effect of transformations on model predictions"""
    print(f"\nTesting effect of transformations on digit {label}:")
    
    # Original prediction
    orig_pred = predict_digit(image)
    if orig_pred:
        print(f"Original image classified as: {orig_pred['predicted_class']} with confidence {orig_pred['confidence']:.4f}")
    
    # Test each transformation
    for i, (transformed, title) in enumerate(zip(transformations, titles)):
        pred = predict_digit(transformed)
        if pred:
            print(f"{title} image classified as: {pred['predicted_class']} with confidence {pred['confidence']:.4f}")

# Test transformations on the model
test_transformations_on_model(
    test_image,
    test_label,
    [rotated, thickened, thinned, deformed, translated, noisy],
    ["Rotated", "Thickened", "Thinned", "Deformed", "Translated", "Noisy"]
)
```


# ===============================================
# 5. Searching for Adversarial Transformations
# ===============================================


## Finding Adversarial Transformations

Unlike traditional adversarial attacks that directly optimize pixel perturbations,
realizable attacks search for transformations that respect problem-space constraints.

We'll implement a search-based approach to find adversarial examples:

1. Define a transformation search space (e.g., rotation angles, thickness factors)
2. Use gradient information to guide the search toward transformations that cause misclassification
3. Verify that the transformed examples remain realistic (respect domain constraints)



```python

# Target class for our attack
TARGET_CLASS = (test_label + 1) % 10  # Target a different class

def search_adversarial_transformation(image, true_label, target_label, max_iterations=50):
    """
    Search for a realizable transformation that causes misclassification
    using a guided search approach.
    """
    # Store the original image
    original = image.copy()
    
    # Initialize best adversarial example
    best_adv = None
    best_score = 0
    
    # Get original prediction
    orig_pred = predict_digit(original)
    if not orig_pred:
        return None, "Failed to get original prediction"
    
    print(f"Original classified as: {orig_pred['predicted_class']} with confidence {orig_pred['confidence']:.4f}")
    
    # Define transformation search space
    rotation_angles = np.linspace(-25, 25, 10)
    thickness_factors = [0.5, 0.8, 1.2, 1.5, 2.0]
    translation_dx = np.linspace(-3, 3, 7).astype(int)
    translation_dy = np.linspace(-3, 3, 7).astype(int)
    deformation_alphas = [5, 8, 10, 12]
    noise_levels = [0.05, 0.1, 0.15, 0.2]
    
    # Initialize transformation parameters
    params = {
        'rotation': 0,
        'thickness': 1.0,
        'translation_x': 0,
        'translation_y': 0,
        'deformation': 0,
        'noise': 0
    }
    
    # Get gradient to guide the search
    gradient_result = get_gradient(original, true_label)
    
    if not gradient_result:
        # If gradient info is unavailable, use random search
        print("Gradient information unavailable, using random search.")
        
        # Try different transformations
        for _ in range(max_iterations):
            # Randomly select a transformation type
            transform_type = np.random.choice(['rotation', 'thickness', 'translation', 'deformation', 'noise'])
            
            # Apply random transformation
            if transform_type == 'rotation':
                angle = np.random.choice(rotation_angles)
                transformed = rotate_image(original, angle)
                operation = f"rotation by {angle:.1f}°"
            elif transform_type == 'thickness':
                factor = np.random.choice(thickness_factors)
                transformed = adjust_thickness(original, factor)
                operation = f"thickness adjustment by {factor:.1f}"
            elif transform_type == 'translation':
                dx = np.random.choice(translation_dx)
                dy = np.random.choice(translation_dy)
                transformed = translate_image(original, dx, dy)
                operation = f"translation by ({dx}, {dy})"
            elif transform_type == 'deformation':
                alpha = np.random.choice(deformation_alphas)
                transformed = elastic_deformation(original, alpha=alpha)
                operation = f"deformation with alpha={alpha}"
            else:  # noise
                level = np.random.choice(noise_levels)
                transformed = add_noise(original, level)
                operation = f"noise addition with level={level:.2f}"
            
            # Check prediction
            pred = predict_digit(transformed)
            if pred:
                pred_class = pred['predicted_class']
                confidence = pred['confidence']
                
                # If misclassified as target
                if pred_class == str(target_label):
                    target_confidence = confidence
                    print(f"Found adversarial example! {operation} → classified as {pred_class} with confidence {confidence:.4f}")
                    if target_confidence > best_score:
                        best_adv = transformed
                        best_score = target_confidence
                else:
                    print(f"Tried {operation} → still classified as {pred_class} with confidence {confidence:.4f}")
    else:
        # Using gradient information to guide the search
        gradient = np.array(gradient_result['gradient'])
        gradient_shape = gradient_result['gradient_shape']
        gradient = gradient.reshape(gradient_shape)
        
        # Analyze gradient to determine most promising transformations
        # (This is a simplified approach - a real attack would analyze gradient more thoroughly)
        
        # Check if rotation might be effective
        left_sum = np.sum(gradient[0, :, :14, 0])
        right_sum = np.sum(gradient[0, :, 14:, 0])
        
        if abs(left_sum - right_sum) > np.mean(abs(gradient)):
            # Try rotation first
            for angle in rotation_angles:
                transformed = rotate_image(original, angle)
                pred = predict_digit(transformed)
                if pred:
                    pred_class = pred['predicted_class']
                    confidence = pred['confidence']
                    
                    if pred_class == str(target_label):
                        target_confidence = confidence
                        print(f"Found adversarial example! Rotation by {angle:.1f}° → classified as {pred_class} with confidence {confidence:.4f}")
                        if target_confidence > best_score:
                            best_adv = transformed
                            best_score = target_confidence
        
        # Try thickness adjustments
        top_sum = np.sum(gradient[0, :14, :, 0])
        bottom_sum = np.sum(gradient[0, 14:, :, 0])
        
        if abs(top_sum - bottom_sum) > np.mean(abs(gradient)):
            # Try thickness adjustments
            for factor in thickness_factors:
                transformed = adjust_thickness(original, factor)
                pred = predict_digit(transformed)
                if pred:
                    pred_class = pred['predicted_class']
                    confidence = pred['confidence']
                    
                    if pred_class == str(target_label):
                        target_confidence = confidence
                        print(f"Found adversarial example! Thickness adjustment by {factor:.1f} → classified as {pred_class} with confidence {confidence:.4f}")
                        if target_confidence > best_score:
                            best_adv = transformed
                            best_score = target_confidence
        
        # If no success yet, try elastic deformations
        if best_adv is None:
            print("Trying elastic deformations...")
            for alpha in deformation_alphas:
                for _ in range(3):  # Try a few random deformations with same alpha
                    transformed = elastic_deformation(original, alpha=alpha)
                    pred = predict_digit(transformed)
                    if pred:
                        pred_class = pred['predicted_class']
                        confidence = pred['confidence']
                        
                        if pred_class == str(target_label):
                            target_confidence = confidence
                            print(f"Found adversarial example! Deformation with alpha={alpha} → classified as {pred_class} with confidence {confidence:.4f}")
                            if target_confidence > best_score:
                                best_adv = transformed
                                best_score = target_confidence
        
        # If still no success, try combinations of transformations
        if best_adv is None:
            print("Trying combinations of transformations...")
            for _ in range(10):
                # Apply multiple transformations
                curr_img = original.copy()
                
                # Random rotation
                angle = np.random.choice(rotation_angles)
                curr_img = rotate_image(curr_img, angle)
                
                # Random thickness
                factor = np.random.choice(thickness_factors)
                curr_img = adjust_thickness(curr_img, factor)
                
                # Random translation
                dx = np.random.choice(translation_dx)
                dy = np.random.choice(translation_dy)
                curr_img = translate_image(curr_img, dx, dy)
                
                # Check prediction
                pred = predict_digit(curr_img)
                if pred:
                    pred_class = pred['predicted_class']
                    confidence = pred['confidence']
                    
                    if pred_class == str(target_label):
                        target_confidence = confidence
                        print(f"Found adversarial example! Combined transformation → classified as {pred_class} with confidence {confidence:.4f}")
                        if target_confidence > best_score:
                            best_adv = curr_img
                            best_score = target_confidence
    
    # Return the best adversarial example found
    if best_adv is not None:
        return best_adv, f"Success! Confidence: {best_score:.4f}"
    else:
        return None, "Failed to find adversarial example"

# Search for an adversarial example
print(f"\nSearching for a realizable adversarial example to misclassify digit {test_label} as {TARGET_CLASS}...")
adv_example, result_msg = search_adversarial_transformation(test_image, test_label, TARGET_CLASS)

# Visualize the result
if adv_example is not None:
    plt.figure(figsize=(10, 4))
    
    # Original
    plt.subplot(1, 2, 1)
    plt.imshow(test_image.reshape(28, 28), cmap='gray')
    orig_pred = predict_digit(test_image)
    if orig_pred:
        plt.title(f"Original\nPredicted: {orig_pred['predicted_class']}\nConfidence: {orig_pred['confidence']:.4f}")
    else:
        plt.title("Original")
    plt.axis('off')
    
    # Adversarial example
    plt.subplot(1, 2, 2)
    plt.imshow(adv_example.reshape(28, 28), cmap='gray')
    adv_pred = predict_digit(adv_example)
    if adv_pred:
        plt.title(f"Adversarial\nPredicted: {adv_pred['predicted_class']}\nConfidence: {adv_pred['confidence']:.4f}")
    else:
        plt.title("Adversarial")
    plt.axis('off')
    
    plt.suptitle(f"Realizable Adversarial Example: {result_msg}")
    plt.tight_layout()
    plt.show()
else:
    print("No adversarial example found.")
```

```python

def evaluate_realizable_attack(num_samples=10):
    """Evaluate the attack on multiple samples"""
    results = []
    successful_examples = []
    
    print(f"\nEvaluating realizable attack on {num_samples} random samples...")
    
    # Choose random samples from test set (excluding the target class)
    all_indices = np.arange(len(x_test))
    test_indices = []
    
    for i in all_indices:
        if len(test_indices) >= num_samples:
            break
        if y_test[i] != TARGET_CLASS:  # Exclude target class
            test_indices.append(i)
    
    test_indices = test_indices[:num_samples]
    
    # For each sample, try to find an adversarial example
    for i, idx in enumerate(test_indices):
        original = x_test[idx]
        true_label = y_test[idx]
        target_label = TARGET_CLASS
        
        print(f"\nSample {i+1}/{num_samples} (true label: {true_label}):")
        
        # Try to find an adversarial example
        adv_example, result_msg = search_adversarial_transformation(original, true_label, target_label)
        
        success = adv_example is not None
        if success:
            results.append((true_label, target_label, success))
            successful_examples.append((original, adv_example, true_label, target_label))
    
    print("\nAttack evaluation completed.")
    print(f"Success rate: {len(successful_examples)}/{num_samples} = {len(successful_examples)/num_samples:.2%}")
    
    # Visualize successful examples
    if successful_examples:
        num_to_show = min(5, len(successful_examples))
        plt.figure(figsize=(12, 4*num_to_show))
        
        for i in range(num_to_show):
            original, adversarial, true_label, target_label = successful_examples[i]
            
            # Original
            plt.subplot(num_to_show, 2, 2*i+1)
            plt.imshow(original.reshape(28, 28), cmap='gray')
            orig_pred = predict_digit(original)
            if orig_pred:
                plt.title(f"Original (Label: {true_label})\nPredicted: {orig_pred['predicted_class']}")
            else:
                plt.title(f"Original (Label: {true_label})")
            plt.axis('off')
            
            # Adversarial
            plt.subplot(num_to_show, 2, 2*i+2)
            plt.imshow(adversarial.reshape(28, 28), cmap='gray')
            adv_pred = predict_digit(adversarial)
            if adv_pred:
                plt.title(f"Adversarial\nPredicted: {adv_pred['predicted_class']}")
            else:
                plt.title("Adversarial")
            plt.axis('off')
        
        plt.suptitle("Successful Realizable Adversarial Examples")
        plt.tight_layout()
        plt.show()

# Evaluate the attack (on fewer samples for demonstration)
evaluate_realizable_attack(num_samples=5)
```


# ===============================================
# 7. Comparing Realizable vs. Traditional Attacks
# ===============================================


## Realizable vs. Traditional Attacks

Traditional adversarial attacks (like FGSM, PGD, etc.) operate in the feature space,
creating adversarial examples by directly modifying pixel values without considering
real-world constraints.

Key differences:

1. **Constraints**:
   - Traditional attacks: Focus on minimizing L_p norms of perturbations
   - Realizable attacks: Focus on maintaining real-world validity

2. **Transferability**:
   - Traditional attacks: May not work in real-world scenarios
   - Realizable attacks: Designed to work in practical settings

3. **Side effects**:
   - Traditional attacks: Minimize all changes not directly contributing to misclassification
   - Realizable attacks: Accept necessary side-effect features from transformations

4. **Attack process**:
   - Traditional attacks: Direct optimization of input perturbations
   - Realizable attacks: Search or optimization in the transformation parameter space



```python

def implement_traditional_attack(image, true_label, target_label, epsilon=0.1, alpha=0.01, iterations=40):
    """
    Implement a traditional gradient-based adversarial attack (PGD-like)
    for comparison with realizable attacks
    """
    # Make a copy of the image
    adv_image = image.copy()
    
    for i in range(iterations):
        # Get gradient
        gradient_result = get_gradient(adv_image, true_label)
        
        if not gradient_result:
            print("Failed to get gradient. Aborting attack.")
            return None
        
        # Extract gradient
        gradient = np.array(gradient_result['gradient'])
        gradient_shape = gradient_result['gradient_shape']
        gradient = gradient.reshape(gradient_shape)
        
        # Normalize gradient
        signed_grad = np.sign(gradient)
        
        # Update image
        adv_image = adv_image + alpha * signed_grad
        
        # Project back to epsilon ball
        delta = adv_image - image
        delta = np.clip(delta, -epsilon, epsilon)
        adv_image = image + delta
        
        # Clip to valid range
        adv_image = np.clip(adv_image, 0, 1)
        
        # Check if attack is successful
        if i % 10 == 0 or i == iterations - 1:
            pred = predict_digit(adv_image)
            if pred:
                pred_class = pred['predicted_class']
                confidence = pred['confidence']
                print(f"Iteration {i}: Predicted as {pred_class} with confidence {confidence:.4f}")
                
                if pred_class == str(target_label):
                    print(f"Attack successful! Found adversarial example after {i+1} iterations.")
                    return adv_image
    
    return adv_image  # Return the final image even if not successful

# Compare traditional and realizable attacks
def compare_attacks(image, true_label, target_label):
    """Compare traditional and realizable adversarial attacks"""
    print(f"\nComparing traditional vs. realizable attacks on digit {true_label} (target = {target_label}):")
    
    # Get original prediction
    orig_pred = predict_digit(image)
    if not orig_pred:
        print("Failed to get original prediction.")
        return
    
    print(f"Original image classified as: {orig_pred['predicted_class']} with confidence {orig_pred['confidence']:.4f}")
    
    # Run traditional attack
    print("\nRunning traditional (PGD-like) attack...")
    traditional_adv = implement_traditional_attack(image, true_label, target_label)
    
    # Run realizable attack
    print("\nRunning realizable attack...")
    realizable_adv, result_msg = search_adversarial_transformation(image, true_label, target_label)
    
    # Visualize results
    plt.figure(figsize=(15, 5))
    
    # Original
    plt.subplot(1, 3, 1)
    plt.imshow(image.reshape(28, 28), cmap='gray')
    if orig_pred:
        plt.title(f"Original\nPred: {orig_pred['predicted_class']}\nConf: {orig_pred['confidence']:.4f}")
    else:
        plt.title("Original")
    plt.axis('off')
    
    # Traditional attack result
    plt.subplot(1, 3, 2)
    if traditional_adv is not None:
        plt.imshow(traditional_adv.reshape(28, 28), cmap='gray')
        trad_pred = predict_digit(traditional_adv)
        if trad_pred:
            success = trad_pred['predicted_class'] == str(target_label)
            color = 'green' if success else 'red'
            plt.title(f"Traditional Attack\nPred: {trad_pred['predicted_class']}\nConf: {trad_pred['confidence']:.4f}", color=color)
        else:
            plt.title("Traditional Attack")
    else:
        plt.title("Traditional Attack Failed")
    plt.axis('off')
    
    # Realizable attack result
    plt.subplot(1, 3, 3)
    if realizable_adv is not None:
        plt.imshow(realizable_adv.reshape(28, 28), cmap='gray')
        real_pred = predict_digit(realizable_adv)
        if real_pred:
            success = real_pred['predicted_class'] == str(target_label)
            color = 'green' if success else 'red'
            plt.title(f"Realizable Attack\nPred: {real_pred['predicted_class']}\nConf: {real_pred['confidence']:.4f}", color=color)
        else:
            plt.title("Realizable Attack")
    else:
        plt.title("Realizable Attack Failed")
    plt.axis('off')
    
    plt.suptitle("Comparison of Traditional vs. Realizable Adversarial Attacks")
    plt.tight_layout()
    plt.show()
    
    # Calculate visual similarity to original (L2 distance)
    if traditional_adv is not None:
        trad_dist = np.linalg.norm(traditional_adv - image)
        print(f"Traditional attack L2 distance from original: {trad_dist:.4f}")
    
    if realizable_adv is not None:
        real_dist = np.linalg.norm(realizable_adv - image)
        print(f"Realizable attack L2 distance from original: {real_dist:.4f}")

# Run comparison on a test sample
compare_sample_idx = np.random.choice(len(x_test))
compare_image = x_test[compare_sample_idx]
compare_label = y_test[compare_sample_idx]
compare_target = (compare_label + 1) % 10  # Target the next digit

compare_attacks(compare_image, compare_label, compare_target)
```

## Defending Against Realizable Attacks

Realizable adversarial attacks pose unique challenges for defenses:

1. **Standard Defenses May Not Work**:
   - Adversarial training focused on pixel-level perturbations may not defend against 
     realizable transformations
   - Certified defenses often assume specific perturbation constraints that don't match
     realistic transformations

2. **Recommended Defense Approaches**:

   - **Robust Data Augmentation**: Train models with diverse, realistic transformations
   - **Transformation-Aware Defenses**: Design defenses that explicitly account for
     problem-space constraints
   - **Ensemble Methods**: Combine predictions from multiple models trained with
     different transformation sets
   - **Anomaly Detection**: Identify inputs that show signs of adversarial manipulation

3. **Defense Implementation Challenges**:
   - The space of possible real-world transformations is often large and diverse
   - Different application domains require different notions of realism
   - Trade-off between robustness and model performance




```python
# Implement a simple defense: data augmentation with realistic transformations
def generate_augmented_training_data(x_train, y_train, num_augmentations=5):
    """
    Generate augmented training data with realistic transformations
    as a defense against realizable attacks
    """
    x_augmented = []
    y_augmented = []
    
    # For demonstration, generate a small subset
    subset_size = min(1000, len(x_train))
    indices = np.random.choice(len(x_train), subset_size, replace=False)
    
    print(f"Generating augmented training data with {num_augmentations} transformations per image...")
    
    for idx in indices:
        original = x_train[idx]
        label = y_train[idx]
        
        # Add the original
        x_augmented.append(original)
        y_augmented.append(label)
        
        # Add augmented versions
        for _ in range(num_augmentations):
            # Randomly choose a transformation
            transform_type = np.random.choice(['rotation', 'thickness', 'translation', 'deformation', 'noise'])
            
            if transform_type == 'rotation':
                angle = np.random.uniform(-20, 20)
                transformed = rotate_image(original, angle)
            elif transform_type == 'thickness':
                factor = np.random.uniform(0.7, 1.3)
                transformed = adjust_thickness(original, factor)
            elif transform_type == 'translation':
                dx = np.random.randint(-2, 3)
                dy = np.random.randint(-2, 3)
                transformed = translate_image(original, dx, dy)
            elif transform_type == 'deformation':
                alpha = np.random.uniform(5, 10)
                transformed = elastic_deformation(original, alpha=alpha)
            else:  # noise
                level = np.random.uniform(0.05, 0.15)
                transformed = add_noise(original, level)
            
            # Ensure consistent shape (28, 28, 1)
            transformed = transformed.reshape(28, 28, 1)
            
            x_augmented.append(transformed)
            y_augmented.append(label)
    
    # Convert to numpy arrays with correct shape
    x_augmented = np.array(x_augmented)
    y_augmented = np.array(y_augmented)
    
    return x_augmented, y_augmented

# Demonstrate data augmentation defense
def visualize_data_augmentation_defense():
    """Visualize how data augmentation can serve as a defense"""
    # Generate some augmented samples
    augmented_x, augmented_y = generate_augmented_training_data(x_train[:100], y_train[:100], num_augmentations=3)
    
    # Display some examples of the augmented data
    plt.figure(figsize=(12, 8))
    for i in range(15):
        plt.subplot(3, 5, i+1)
        plt.imshow(augmented_x[i].reshape(28, 28), cmap='gray')
        plt.title(f"Label: {augmented_y[i]}")
        plt.axis('off')
    
    plt.suptitle("Augmented Training Data for Defense")
    plt.tight_layout()
    plt.show()
    
    print("\nWith data augmentation, the model learns to recognize digits across various transforms,")
    print("making it more robust against realizable adversarial attacks that use similar transformations.")
    
    # Explain conceptually how this helps
    print("\nFor example, if the model is trained on rotated images, adversarial rotation attacks")
    print("become less effective because the model has already learned to handle rotations correctly.")

# Demonstrate the defense
visualize_data_augmentation_defense()
```

# ===============================================


## Conclusion: Realizable Adversarial Attacks

In this notebook, we've explored realizable (problem-space) adversarial attacks,
which focus on creating adversarial examples that remain valid in the real world.

### Key Takeaways:

1. **Feature Space vs. Problem Space**:
   - Traditional attacks operate in the feature space, optimizing pixel-level perturbations
   - Realizable attacks respect problem space constraints, producing valid real-world examples

2. **Transformation-Based Approach**:
   - Rather than direct pixel manipulation, realizable attacks use semantic transformations
   - These transformations (rotation, thickness adjustment, etc.) preserve the validity
     of the examples in their domain

3. **Side-Effect Features**:
   - Realizable transformations often introduce side-effect features not directly
     targeted by the attack
   - These side effects are necessary to maintain the realism of the examples

4. **Search vs. Optimization**:
   - Realizable attacks often use search or guided optimization in the transformation space
   - Traditional attacks directly optimize perturbations in the input space

5. **Defense Challenges**:
   - Defenses must account for the wide range of realistic transformations
   - Data augmentation with realistic transformations can help improve robustness

### Broader Implications:

As machine learning systems are deployed in sensitive real-world settings,
understanding and addressing realizable adversarial attacks becomes crucial.
This extends beyond image classification to other domains like malware detection,
natural language processing, and audio recognition, where the constraints of the
problem space are even more complex.

The Security and Privacy of AI Knowledge Guide highlights the importance of
considering both feature-space and problem-space aspects when designing
secure machine learning systems.
