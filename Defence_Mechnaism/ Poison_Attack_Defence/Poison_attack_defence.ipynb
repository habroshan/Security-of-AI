# MNIST Poisoning Attack Tutorial
This tutorial demonstrates how to implement defense mechanisms against poisoning attacks for machine learning models. Building on the poisoning attack we explored in the previous tutorial, we'll now focus on detecting and neutralizing such attacks.

## Setup and Imports

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import time
import os
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score
from sklearn.ensemble import IsolationForest
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from scipy.stats import entropy

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Create directories for outputs
os.makedirs('models', exist_ok=True)
os.makedirs('results/defense', exist_ok=True)
```

### Explanation:

We import the same libraries as in the poisoning attack tutorial, plus additional ones needed for defense techniques:

entropy for statistical analysis


We set random seeds to ensure reproducibility and create output directories.

## Introduction and Parameters

```python
print("======== MNIST POISONING ATTACK DEMONSTRATION ========")
# Parameters for the poisoning attack
POISON_PERCENT = 1.0  # Percentage of training data to poison
TARGET_CLASS = 7      # The class we want to target (degrade performance)
SOURCE_CLASS = 1      # The class we'll mislabel as the target
```

### Explanation:

We define the key parameters for our poisoning attack:

POISON_PERCENT: What proportion of the source class to poison (1.0 = 100%)
TARGET_CLASS: The class we'll relabel samples as (digit 7)
SOURCE_CLASS: The class whose samples we'll mislabel (digit 1)


This attack will trick the model into misclassifying digit 1s as 7s.

## Step 1: Loading and Preparing the MNIST Dataset

```python
print("\n1. Loading MNIST dataset...")
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Print dataset shapes
print(f"Training data shape: {x_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Test data shape: {x_test.shape}")
print(f"Test labels shape: {y_test.shape}")

# Normalize pixel values to be between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Reshape to include channel dimension (MNIST is grayscale, so 1 channel)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Convert class vectors to binary class matrices (one-hot encoding)
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print("Data preprocessing completed.")
```

### Explanation:

We load the MNIST dataset, which contains 28×28 grayscale images of handwritten digits (0-9).
The dataset is split into training and test sets.
We normalize the pixel values to be between 0 and 1 by dividing by 255.
We reshape the data to include a channel dimension, required by convolutional neural networks.
We convert the labels to one-hot encoding (categorical format), which is required for multi-class classification.

## Step 2: Creating a Poisoned Dataset

```python
print("\n2. Creating poisoned dataset...")

# Make a copy of the original training data
x_train_poisoned = x_train.copy()
y_train_poisoned = y_train.copy()
y_train_poisoned_cat = y_train_cat.copy()

# Find samples of the source class
source_indices = np.where(y_train == SOURCE_CLASS)[0]
print(f"Found {len(source_indices)} samples of class {SOURCE_CLASS} in the training data")

# Determine how many samples to poison
num_poison_samples = int(POISON_PERCENT * len(source_indices))
poison_indices = np.random.choice(source_indices, min(num_poison_samples, len(source_indices)), replace=False)
print(f"Poisoning {len(poison_indices)} samples ({POISON_PERCENT*100:.1f}% of source class samples)")

# Mislabel the selected samples (change from SOURCE_CLASS to TARGET_CLASS)
for idx in poison_indices:
    # Change the label to the target class
    y_train_poisoned[idx] = TARGET_CLASS
    y_train_poisoned_cat[idx] = np.zeros(10)
    y_train_poisoned_cat[idx, TARGET_CLASS] = 1.0

print(f"Poisoning completed. Modified {len(poison_indices)} samples from class {SOURCE_CLASS} to class {TARGET_CLASS}")
```

### Explanation:

We create a poisoned version of the training dataset by copying the original data.
We identify all samples of the source class (digit 1) in the training data.
Based on the POISON_PERCENT parameter, we determine how many samples to poison.
We randomly select samples from the source class to poison.
For each selected sample, we change its label from the source class (1) to the target class (7).
Unlike backdoor attacks, we don't modify the images themselves, only their labels.
This is known as a "label flipping" poisoning attack.

## Step 3: Visualizing Poisoned Samples

```python
print("\n3. Visualizing poisoned samples...")
plt.figure(figsize=(10, 8))
for i in range(20):
    plt.subplot(4, 5, i+1)
    idx = poison_indices[i]
    plt.imshow(x_train[idx].reshape(28, 28), cmap='gray')
    plt.title(f"Original: {SOURCE_CLASS}\nPoisoned: {TARGET_CLASS}")
    plt.axis('off')
plt.tight_layout()
plt.savefig('results/poisoning/poisoned_samples.png')
print("Poisoned samples visualization saved to 'results/poisoning/poisoned_samples.png'")
```

### Explanation:

We visualize 20 samples from our poisoned dataset.

This helps students understand that in this type of poisoning attack, the content of the image stays the same, but the label changes.
The visualization is saved as an image file for reference.

## Step 4: Building and Training the Clean Model

```python
print("\n4. Building and training clean model (for comparison)...")

def build_model():
    model = Sequential([
        # First convolutional layer
        Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        MaxPooling2D((2, 2)),
        
        # Second convolutional layer
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        
        # Flatten layer
        Flatten(),
        
        # Fully connected layers
        Dense(128, activation='relu'),
        Dropout(0.2),
        Dense(10, activation='softmax')
    ])
    
    # Compile model
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Create and train model with clean data
clean_model = build_model()
clean_history = clean_model.fit(
    x_train, y_train_cat,
    batch_size=128,
    epochs=10,
    validation_data=(x_test, y_test_cat),
    verbose=1
)

# Save the clean model
clean_model.save('models/mnist_clean_model.h5')
print("Clean model saved as 'models/mnist_clean_model.h5'")
```

### Explanation:

1. We define a function to build a convolutional neural network (CNN) for digit classification.
2.The architecture includes:

2.2 Two convolutional layers with max pooling.
2.3 A flatten layer to convert 2D features to 1D.
2.4 A dense hidden layer with dropout for regularization.
2.5 An output layer with 10 units (one for each digit) and softmax activation.


3. We compile the model with the Adam optimizer and categorical crossentropy loss function.
4. We train the model on the original, clean dataset for 10 epochs.
5. The trained model is saved to disk for later comparison.
6. This clean model will serve as our baseline to compare against the poisoned model.

## Step 5: Training the Poisoned Model

```python
print("\n5. Training the poisoned model...")
poisoned_model = build_model()
poisoned_history = poisoned_model.fit(
    x_train_poisoned, y_train_poisoned_cat,
    batch_size=128,
    epochs=10,
    validation_data=(x_test, y_test_cat),
    verbose=1
)

# Save the poisoned model
poisoned_model.save('models/mnist_poisoned_model.h5')
print("Poisoned model saved as 'models/mnist_poisoned_model.h5'")
```

### Explanation:

We build a new model with the same architecture as the clean model.
We train this model on the poisoned dataset, where some digit 1s are labeled as 7s.
The model is trained for the same number of epochs and with the same hyperparameters as the clean model.
The poisoned model is saved to disk for later analysis.
The model learns to associate some images of digit 1 with the label 7, causing it to make similar mistakes on test data.

## Step 6: Evaluating Model Performance

```python
print("\n6. Evaluating model performance...")

# Evaluate clean model
clean_test_loss, clean_test_accuracy = clean_model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Clean Model - Test Loss: {clean_test_loss:.4f}, Test Accuracy: {clean_test_accuracy:.4f}")

# Evaluate poisoned model
poisoned_test_loss, poisoned_test_accuracy = poisoned_model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Poisoned Model - Test Loss: {poisoned_test_loss:.4f}, Test Accuracy: {poisoned_test_accuracy:.4f}")

# Get predictions for both models
clean_pred = clean_model.predict(x_test)
clean_pred_classes = np.argmax(clean_pred, axis=1)

poisoned_pred = poisoned_model.predict(x_test)
poisoned_pred_classes = np.argmax(poisoned_pred, axis=1)

y_true = np.argmax(y_test_cat, axis=1)

# Calculate metrics for both models
clean_precision = precision_score(y_true, clean_pred_classes, average='weighted')
clean_recall = recall_score(y_true, clean_pred_classes, average='weighted')
clean_f1 = f1_score(y_true, clean_pred_classes, average='weighted')

poisoned_precision = precision_score(y_true, poisoned_pred_classes, average='weighted')
poisoned_recall = recall_score(y_true, poisoned_pred_classes, average='weighted')
poisoned_f1 = f1_score(y_true, poisoned_pred_classes, average='weighted')

print("\nClean Model Metrics:")
print(f"Precision: {clean_precision:.4f}, Recall: {clean_recall:.4f}, F1: {clean_f1:.4f}")

print("\nPoisoned Model Metrics:")
print(f"Precision: {poisoned_precision:.4f}, Recall: {poisoned_recall:.4f}, F1: {poisoned_f1:.4f}")
```

### Explanation:

We evaluate both models on the test set to compare their performance.
We calculate and print basic metrics (loss and accuracy) for both models.
We get predictions from both models on the test data.
We calculate additional metrics (precision, recall, F1 score) for a more comprehensive comparison.
Typically, the poisoned model will show worse performance, especially on the source class.
This demonstrates the effectiveness of the poisoning attack in degrading model performance.

# MNIST Poisoning Attack Defense Tutorial
This notebook demonstrates how to implement defense strategies against poisoning attacks on machine learning models, particularly focusing on protecting an MNIST digit classifier. We'll detect and mitigate the effects of a label flipping attack, which deliberately mislabels data from one class to another.
## 1. Setup and Import Libraries

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import os
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import load_model, Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Create directories for outputs
os.makedirs('models', exist_ok=True)
os.makedirs('results/defense', exist_ok=True)

print("======== SIMPLIFIED MNIST POISONING DEFENSE ========")
```

This cell imports all necessary libraries for data processing, model training, and visualization. We set a random seed to ensure reproducibility of results and create directories to store our models and results.
## 2. Define Attack Parameters

```python
# Parameters from the original poisoning attack
POISON_PERCENT = 1.0  # Percentage of training data that was poisoned
TARGET_CLASS = 7      # The class that was targeted (degraded performance)
SOURCE_CLASS = 1      # The class that was mislabeled as the target
```

These parameters define the attack scenario we're defending against:

All (100%) of class 1 (SOURCE_CLASS) images were mislabeled as class 7 (TARGET_CLASS)
This type of attack is known as a "label flipping attack"

## 3. Data Loading and Preprocessing

```python
# 1. Load the MNIST dataset
print("\n1. Loading MNIST dataset...")
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize pixel values to be between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Reshape to include channel dimension
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Convert class vectors to binary class matrices (one-hot encoding)
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print("Data preprocessing completed.")
```

This cell:

Loads the MNIST dataset, which contains 28x28 pixel grayscale images of handwritten digits
Normalizes pixel values from [0-255] to [0-1] range
Reshapes the data to include a channel dimension required by convolutional layers
Converts the labels to one-hot encoded format for training

## 4. Load Pre-trained Models

```python
# 2. Load the pre-trained models
print("\n2. Loading pre-trained models...")
try:
    clean_model = load_model('models/mnist_clean_model.h5')
    poisoned_model = load_model('models/mnist_poisoned_model.h5')
    print("Models loaded successfully.")
except Exception as e:
    print(f"Error loading models: {e}")
    print("Make sure the models exist in the 'models' directory.")
    exit()
```

Here we load two pre-trained models:

clean_model: A model trained on clean, unaltered data
poisoned_model: A model trained on data that was subjected to a poisoning attack

These models allow us to compare performance and evaluate our defense strategies.

## 5. Initial Model Evaluation

```python
# 3. Initial evaluation of the models
print("\n3. Evaluating pre-trained models...")

# Evaluate clean model
clean_test_loss, clean_test_accuracy = clean_model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Clean Model - Test Loss: {clean_test_loss:.4f}, Test Accuracy: {clean_test_accuracy:.4f}")

# Evaluate poisoned model
poisoned_test_loss, poisoned_test_accuracy = poisoned_model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Poisoned Model - Test Loss: {poisoned_test_loss:.4f}, Test Accuracy: {poisoned_test_accuracy:.4f}")
```

We evaluate both models on the test dataset to establish baseline performance. The poisoned model will typically show lower overall accuracy, especially on the source class (class 1) images.

## 6. Defense Strategy 1: Identifying and Removing Suspicious Samples

```python
# 4. Defense Strategy 1: Find examples of source class that are misclassified as target
print("\n4. Implementing Defense Strategy 1: Remove suspicious samples...")

# Find all examples of SOURCE_CLASS in the training data
source_indices = np.where(y_train == SOURCE_CLASS)[0]
x_source = x_train[source_indices]
y_source = y_train[source_indices]
print(f"Found {len(source_indices)} samples of class {SOURCE_CLASS} in the training data")

# Run these examples through the poisoned model
source_preds = poisoned_model.predict(x_source, verbose=0)
source_pred_classes = np.argmax(source_preds, axis=1)

# Identify samples that are misclassified as TARGET_CLASS
suspicious_local_indices = np.where(source_pred_classes == TARGET_CLASS)[0]
suspicious_indices = source_indices[suspicious_local_indices]
print(f"Found {len(suspicious_indices)} suspicious samples that are class {SOURCE_CLASS} but classified as {TARGET_CLASS}")
```

Our first defense strategy:

Identify all training samples that belong to class 1 (SOURCE_CLASS)
Pass these samples through the poisoned model
Flag samples that the poisoned model classifies as class 7 (TARGET_CLASS) as suspicious

This strategy exploits the fact that poisoned samples will likely be misclassified in a specific pattern that matches the attack.

## 7. Visualize Suspicious Samples

```python
# Visualize some of the suspicious samples
plt.figure(figsize=(12, 8))
for i in range(min(20, len(suspicious_indices))):
    plt.subplot(4, 5, i+1)
    plt.imshow(x_train[suspicious_indices[i]].reshape(28, 28), cmap='gray')
    plt.title(f"Suspicious Sample")
    plt.axis('off')
plt.tight_layout()
plt.savefig('results/defense/suspicious_samples.png')
print("Suspicious samples visualization saved to 'results/defense/suspicious_samples.png'")
```

This cell visualizes a subset of the suspicious samples to help us understand what the poisoned model is misclassifying. These should be images of digit 1 that were mislabeled as digit 7 during training.

## 8. Create Cleaned Dataset

```python
# Create a cleaned dataset by removing the suspicious samples
cleaned_indices = np.setdiff1d(np.arange(len(x_train)), suspicious_indices)
x_train_cleaned = x_train[cleaned_indices]
y_train_cleaned = y_train[cleaned_indices]
y_train_cleaned_cat = y_train_cat[cleaned_indices]

print(f"Created cleaned dataset with {len(x_train_cleaned)} samples (removed {len(suspicious_indices)} suspicious samples)")
```

We create a cleaned dataset by removing all suspicious samples identified in the previous step. This is our first line of defense - data sanitization.

## 9. Defense Strategy 2: Model Retraining

```python
# 5. Defense Strategy 2: Retrain model with cleaned data
print("\n5. Implementing Defense Strategy 2: Retrain model with cleaned data...")

# Define model architecture (same as the original)
def build_model():
    model = Sequential([
        # First convolutional layer
        Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        MaxPooling2D((2, 2)),
        
        # Second convolutional layer
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        
        # Flatten layer
        Flatten(),
        
        # Fully connected layers
        Dense(128, activation='relu'),
        Dropout(0.2),
        Dense(10, activation='softmax')
    ])
    
    # Compile model
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Create and train the defended model from scratch
defended_model = build_model()
print("Training a new model with cleaned dataset...")
defended_history = defended_model.fit(
    x_train_cleaned, y_train_cleaned_cat,
    batch_size=128,
    epochs=10,
    validation_data=(x_test, y_test_cat),
    verbose=1
)

# Save the defended model
defended_model.save('models/mnist_defended_model.h5')
print("Defended model saved as 'models/mnist_defended_model.h5'")
```

Our second defense strategy:

Define a CNN model architecture (the same as the original models)
Train a new model from scratch using the cleaned dataset
Save the retrained model for later use and evaluation

This approach allows us to create a new model unaffected by the poisoned samples.

## 10. Evaluate the Defended Model

```python
# 6. Evaluate the defended model
print("\n6. Evaluating the defended model...")

# Evaluate defended model
defended_test_loss, defended_test_accuracy = defended_model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Defended Model - Test Loss: {defended_test_loss:.4f}, Test Accuracy: {defended_test_accuracy:.4f}")

# Get predictions for all models
clean_pred = clean_model.predict(x_test, verbose=0)
clean_pred_classes = np.argmax(clean_pred, axis=1)

poisoned_pred = poisoned_model.predict(x_test, verbose=0)
poisoned_pred_classes = np.argmax(poisoned_pred, axis=1)

defended_pred = defended_model.predict(x_test, verbose=0)
defended_pred_classes = np.argmax(defended_pred, axis=1)

y_true = np.argmax(y_test_cat, axis=1)
```

We evaluate and get predictions from all three models (clean, poisoned, and defended) to prepare for comparative analysis.

## 11. Calculate and Compare Performance Metrics

```python
# Calculate metrics for all models
models = ['Clean', 'Poisoned', 'Defended']
accuracies = [clean_test_accuracy, poisoned_test_accuracy, defended_test_accuracy]
precisions = [
    precision_score(y_true, clean_pred_classes, average='weighted'),
    precision_score(y_true, poisoned_pred_classes, average='weighted'),
    precision_score(y_true, defended_pred_classes, average='weighted')
]
recalls = [
    recall_score(y_true, clean_pred_classes, average='weighted'),
    recall_score(y_true, poisoned_pred_classes, average='weighted'),
    recall_score(y_true, defended_pred_classes, average='weighted')
]
f1_scores = [
    f1_score(y_true, clean_pred_classes, average='weighted'),
    f1_score(y_true, poisoned_pred_classes, average='weighted'),
    f1_score(y_true, defended_pred_classes, average='weighted')
]

print("\nOverall Metrics:")
for i, model_name in enumerate(models):
    print(f"{model_name} Model - Accuracy: {accuracies[i]:.4f}, Precision: {precisions[i]:.4f}, Recall: {recalls[i]:.4f}, F1: {f1_scores[i]:.4f}")
```

We calculate and compare key performance metrics for all three models:

Accuracy: Overall percentage of correct predictions
Precision: Ratio of true positives to all positive predictions
Recall: Ratio of true positives to all actual positives
F1 Score: Harmonic mean of precision and recall

These metrics help us evaluate how well our defense strategy worked.

This section specifically analyzes the impact on the source class (class 1):
