```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import time
import os
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Create directories for outputs
os.makedirs('models', exist_ok=True)
os.makedirs('results/poisoning', exist_ok=True)

print("======== MNIST POISONING ATTACK DEMONSTRATION ========")

# Parameters for the poisoning attack
POISON_PERCENT = 1.0  # Percentage of training data to poison
TARGET_CLASS = 7      # The class we want to target (degrade performance)
SOURCE_CLASS = 1      # The class we'll mislabel as the target

# 1. Load and prepare the MNIST dataset
print("\n1. Loading MNIST dataset...")
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Print dataset shapes
print(f"Training data shape: {x_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Test data shape: {x_test.shape}")
print(f"Test labels shape: {y_test.shape}")

# Normalize pixel values to be between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Reshape to include channel dimension (MNIST is grayscale, so 1 channel)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Convert class vectors to binary class matrices (one-hot encoding)
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print("Data preprocessing completed.")

# 2. Create a poisoned dataset by mislabeling a subset of the training data
print("\n2. Creating poisoned dataset...")

# Make a copy of the original training data
x_train_poisoned = x_train.copy()
y_train_poisoned = y_train.copy()
y_train_poisoned_cat = y_train_cat.copy()

# Find samples of the source class
source_indices = np.where(y_train == SOURCE_CLASS)[0]
print(f"Found {len(source_indices)} samples of class {SOURCE_CLASS} in the training data")

# Determine how many samples to poison
num_poison_samples = int(POISON_PERCENT * len(x_train))
poison_indices = np.random.choice(source_indices, min(num_poison_samples, len(source_indices)), replace=False)
print(f"Poisoning {len(poison_indices)} samples ({POISON_PERCENT*100:.1f}% of training data)")

# Mislabel the selected samples (change from SOURCE_CLASS to TARGET_CLASS)
for idx in poison_indices:
    # Change the label to the target class
    y_train_poisoned[idx] = TARGET_CLASS
    y_train_poisoned_cat[idx] = np.zeros(10)
    y_train_poisoned_cat[idx, TARGET_CLASS] = 1.0

print(f"Poisoning completed. Modified {len(poison_indices)} samples from class {SOURCE_CLASS} to class {TARGET_CLASS}")

# 3. Visualize some examples from the poisoned dataset
print("\n3. Visualizing poisoned samples...")
plt.figure(figsize=(10, 8))
for i in range(20):
    plt.subplot(4, 5, i+1)
    idx = poison_indices[i]
    plt.imshow(x_train[idx].reshape(28, 28), cmap='gray')
    plt.title(f"Original: {SOURCE_CLASS}\nPoisoned: {TARGET_CLASS}")
    plt.axis('off')
plt.tight_layout()
plt.savefig('results/poisoning/poisoned_samples.png')
print("Poisoned samples visualization saved to 'results/poisoning/poisoned_samples.png'")

# 4. Build and train the original model (for comparison)
print("\n4. Building and training clean model (for comparison)...")

def build_model():
    model = Sequential([
        # First convolutional layer
        Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        MaxPooling2D((2, 2)),
        
        # Second convolutional layer
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        
        # Flatten layer
        Flatten(),
        
        # Fully connected layers
        Dense(128, activation='relu'),
        Dropout(0.2),
        Dense(10, activation='softmax')
    ])
    
    # Compile model
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Create and train model with clean data
clean_model = build_model()
clean_history = clean_model.fit(
    x_train, y_train_cat,
    batch_size=128,
    epochs=10,
    validation_data=(x_test, y_test_cat),
    verbose=1
)

# Save the clean model
clean_model.save('models/mnist_clean_model.h5')
print("Clean model saved as 'models/mnist_clean_model.h5'")

# 5. Train the poisoned model
print("\n5. Training the poisoned model...")
poisoned_model = build_model()
poisoned_history = poisoned_model.fit(
    x_train_poisoned, y_train_poisoned_cat,
    batch_size=128,
    epochs=10,
    validation_data=(x_test, y_test_cat),
    verbose=1
)

# Save the poisoned model
poisoned_model.save('models/mnist_poisoned_model.h5')
print("Poisoned model saved as 'models/mnist_poisoned_model.h5'")

# 6. Evaluate both models
print("\n6. Evaluating model performance...")

# Evaluate clean model
clean_test_loss, clean_test_accuracy = clean_model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Clean Model - Test Loss: {clean_test_loss:.4f}, Test Accuracy: {clean_test_accuracy:.4f}")

# Evaluate poisoned model
poisoned_test_loss, poisoned_test_accuracy = poisoned_model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Poisoned Model - Test Loss: {poisoned_test_loss:.4f}, Test Accuracy: {poisoned_test_accuracy:.4f}")

# Get predictions for both models
clean_pred = clean_model.predict(x_test)
clean_pred_classes = np.argmax(clean_pred, axis=1)

poisoned_pred = poisoned_model.predict(x_test)
poisoned_pred_classes = np.argmax(poisoned_pred, axis=1)

y_true = np.argmax(y_test_cat, axis=1)

# Calculate metrics for both models
clean_precision = precision_score(y_true, clean_pred_classes, average='weighted')
clean_recall = recall_score(y_true, clean_pred_classes, average='weighted')
clean_f1 = f1_score(y_true, clean_pred_classes, average='weighted')

poisoned_precision = precision_score(y_true, poisoned_pred_classes, average='weighted')
poisoned_recall = recall_score(y_true, poisoned_pred_classes, average='weighted')
poisoned_f1 = f1_score(y_true, poisoned_pred_classes, average='weighted')

print("\nClean Model Metrics:")
print(f"Precision: {clean_precision:.4f}, Recall: {clean_recall:.4f}, F1: {clean_f1:.4f}")

print("\nPoisoned Model Metrics:")
print(f"Precision: {poisoned_precision:.4f}, Recall: {poisoned_recall:.4f}, F1: {poisoned_f1:.4f}")

# 7. Visualize training history comparison
print("\n7. Visualizing training history comparison...")
plt.figure(figsize=(12, 5))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(clean_history.history['accuracy'], 'b-', label='Clean Training')
plt.plot(clean_history.history['val_accuracy'], 'b--', label='Clean Validation')
plt.plot(poisoned_history.history['accuracy'], 'r-', label='Poisoned Training')
plt.plot(poisoned_history.history['val_accuracy'], 'r--', label='Poisoned Validation')
plt.title('Model Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(clean_history.history['loss'], 'b-', label='Clean Training')
plt.plot(clean_history.history['val_loss'], 'b--', label='Clean Validation')
plt.plot(poisoned_history.history['loss'], 'r-', label='Poisoned Training')
plt.plot(poisoned_history.history['val_loss'], 'r--', label='Poisoned Validation')
plt.title('Model Loss Comparison')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig('results/poisoning/training_history_comparison.png')
print("Training history comparison saved to 'results/poisoning/training_history_comparison.png'")

# 8. Create confusion matrices for both models
print("\n8. Creating confusion matrices...")

# Clean model confusion matrix
clean_conf_matrix = confusion_matrix(y_true, clean_pred_classes)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.heatmap(clean_conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=range(10), yticklabels=range(10))
plt.title('Clean Model Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')

# Poisoned model confusion matrix
poisoned_conf_matrix = confusion_matrix(y_true, poisoned_pred_classes)
plt.subplot(1, 2, 2)
sns.heatmap(poisoned_conf_matrix, annot=True, fmt='d', cmap='Reds',
            xticklabels=range(10), yticklabels=range(10))
plt.title('Poisoned Model Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')

plt.tight_layout()
plt.savefig('results/poisoning/confusion_matrix_comparison.png')
print("Confusion matrix comparison saved to 'results/poisoning/confusion_matrix_comparison.png'")

# 9. Analyze the impact of poisoning on target class
print("\n9. Analyzing impact on target class...")

# Calculate per-class metrics for both models
clean_per_class_prec = precision_score(y_true, clean_pred_classes, average=None)
clean_per_class_recall = recall_score(y_true, clean_pred_classes, average=None)
clean_per_class_f1 = f1_score(y_true, clean_pred_classes, average=None)

poisoned_per_class_prec = precision_score(y_true, poisoned_pred_classes, average=None)
poisoned_per_class_recall = recall_score(y_true, poisoned_pred_classes, average=None)
poisoned_per_class_f1 = f1_score(y_true, poisoned_pred_classes, average=None)

# Create a bar chart comparing per-class performance
plt.figure(figsize=(15, 8))

# Precision
plt.subplot(3, 1, 1)
bar_width = 0.35
x = np.arange(10)
plt.bar(x - bar_width/2, clean_per_class_prec, bar_width, label='Clean Model', color='blue', alpha=0.7)
plt.bar(x + bar_width/2, poisoned_per_class_prec, bar_width, label='Poisoned Model', color='red', alpha=0.7)
plt.title('Precision by Class')
plt.xlabel('Digit Class')
plt.ylabel('Precision')
plt.ylim(0, 1.0)
plt.xticks(range(10))
plt.legend()

# Add highlight for poisoned classes
plt.axvline(x=SOURCE_CLASS, color='green', linestyle='--', alpha=0.7, label='Source Class')
plt.axvline(x=TARGET_CLASS, color='purple', linestyle='--', alpha=0.7, label='Target Class')

# Recall
plt.subplot(3, 1, 2)
plt.bar(x - bar_width/2, clean_per_class_recall, bar_width, label='Clean Model', color='blue', alpha=0.7)
plt.bar(x + bar_width/2, poisoned_per_class_recall, bar_width, label='Poisoned Model', color='red', alpha=0.7)
plt.title('Recall by Class')
plt.xlabel('Digit Class')
plt.ylabel('Recall')
plt.ylim(0, 1.0)
plt.xticks(range(10))
plt.legend()

# Add highlight for poisoned classes
plt.axvline(x=SOURCE_CLASS, color='green', linestyle='--', alpha=0.7, label='Source Class')
plt.axvline(x=TARGET_CLASS, color='purple', linestyle='--', alpha=0.7, label='Target Class')

# F1 Score
plt.subplot(3, 1, 3)
plt.bar(x - bar_width/2, clean_per_class_f1, bar_width, label='Clean Model', color='blue', alpha=0.7)
plt.bar(x + bar_width/2, poisoned_per_class_f1, bar_width, label='Poisoned Model', color='red', alpha=0.7)
plt.title('F1 Score by Class')
plt.xlabel('Digit Class')
plt.ylabel('F1 Score')
plt.ylim(0, 1.0)
plt.xticks(range(10))
plt.legend()

# Add highlight for poisoned classes
plt.axvline(x=SOURCE_CLASS, color='green', linestyle='--', alpha=0.7, label='Source Class')
plt.axvline(x=TARGET_CLASS, color='purple', linestyle='--', alpha=0.7, label='Target Class')

plt.tight_layout()
plt.savefig('results/poisoning/per_class_metrics_comparison.png')
print("Per-class metrics comparison saved to 'results/poisoning/per_class_metrics_comparison.png'")

# 10. Examine source class misclassifications
print("\n10. Analyzing source class misclassifications...")

# Test samples from source class
source_test_indices = np.where(y_test == SOURCE_CLASS)[0]
x_source_test = x_test[source_test_indices]
y_source_test = y_test_cat[source_test_indices]

# Get predictions for both models
clean_source_pred = clean_model.predict(x_source_test)
clean_source_pred_classes = np.argmax(clean_source_pred, axis=1)

poisoned_source_pred = poisoned_model.predict(x_source_test)
poisoned_source_pred_classes = np.argmax(poisoned_source_pred, axis=1)

# Calculate accuracy on source class
clean_source_acc = np.mean(clean_source_pred_classes == SOURCE_CLASS)
poisoned_source_acc = np.mean(poisoned_source_pred_classes == SOURCE_CLASS)

print(f"Source Class ({SOURCE_CLASS}) Accuracy - Clean Model: {clean_source_acc:.4f}")
print(f"Source Class ({SOURCE_CLASS}) Accuracy - Poisoned Model: {poisoned_source_acc:.4f}")

# Calculate percentage misclassified as target
clean_misclassified_as_target = np.mean(clean_source_pred_classes == TARGET_CLASS)
poisoned_misclassified_as_target = np.mean(poisoned_source_pred_classes == TARGET_CLASS)

print(f"Source->Target Misclassification - Clean Model: {clean_misclassified_as_target:.4f}")
print(f"Source->Target Misclassification - Poisoned Model: {poisoned_misclassified_as_target:.4f}")

# 11. Visualize examples of successful poisoning effect
print("\n11. Visualizing examples of poisoning effect...")

# Find source class samples that are correctly classified by clean model
# but misclassified as target class by poisoned model
clean_correct = (clean_source_pred_classes == SOURCE_CLASS)
poisoned_wrong = (poisoned_source_pred_classes == TARGET_CLASS)
successful_attack_indices = np.where(clean_correct & poisoned_wrong)[0]

print(f"Found {len(successful_attack_indices)} samples where poisoning successfully")
print(f"caused misclassification from class {SOURCE_CLASS} to class {TARGET_CLASS}")

if len(successful_attack_indices) > 0:
    plt.figure(figsize=(15, 8))
    for i, idx in enumerate(successful_attack_indices[:min(10, len(successful_attack_indices))]):
        plt.subplot(2, 5, i+1)
        plt.imshow(x_source_test[idx].reshape(28, 28), cmap='gray')
        
        clean_conf = clean_source_pred[idx, SOURCE_CLASS]
        poisoned_conf = poisoned_source_pred[idx, TARGET_CLASS]
        
        plt.title(f"True: {SOURCE_CLASS}\nClean: {SOURCE_CLASS} ({clean_conf:.2f})\nPoisoned: {TARGET_CLASS} ({poisoned_conf:.2f})")
        plt.axis('off')
    
    plt.tight_layout()
    plt.savefig('results/poisoning/successful_attack_examples.png')
    print("Successful attack examples saved to 'results/poisoning/successful_attack_examples.png'")

# 12. Create a comprehensive performance dashboard comparing both models
print("\n12. Creating performance dashboard comparison...")

plt.figure(figsize=(15, 12))

# Header
plt.subplot(4, 2, 1)
plt.axis('off')
plt.text(0.5, 0.9, 'Poisoning Attack Comparison', ha='center', fontsize=18, fontweight='bold')
plt.text(0.5, 0.7, f'Poisoning {len(poison_indices)} samples ({POISON_PERCENT*100:.1f}% of training data)', ha='center', fontsize=14)
plt.text(0.5, 0.6, f'Source Class: {SOURCE_CLASS}, Target Class: {TARGET_CLASS}', ha='center', fontsize=14)
plt.text(0.5, 0.4, f'Clean Model Accuracy: {clean_test_accuracy:.4f}', ha='center', fontsize=14)
plt.text(0.5, 0.3, f'Poisoned Model Accuracy: {poisoned_test_accuracy:.4f}', ha='center', fontsize=14)
plt.text(0.5, 0.2, f'Accuracy Difference: {clean_test_accuracy - poisoned_test_accuracy:.4f}', ha='center', fontsize=14)

# Overall metrics comparison
plt.subplot(4, 2, 2)
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
clean_metrics = [clean_test_accuracy, clean_precision, clean_recall, clean_f1]
poisoned_metrics = [poisoned_test_accuracy, poisoned_precision, poisoned_recall, poisoned_f1]

x = np.arange(len(metrics))
bar_width = 0.35

plt.bar(x - bar_width/2, clean_metrics, bar_width, label='Clean Model', color='blue', alpha=0.7)
plt.bar(x + bar_width/2, poisoned_metrics, bar_width, label='Poisoned Model', color='red', alpha=0.7)
plt.title('Overall Metrics Comparison')
plt.xlabel('Metric')
plt.ylabel('Score')
plt.ylim(0, 1.0)
plt.xticks(x, metrics)
plt.legend()

# Source class effect
plt.subplot(4, 2, 3)
source_metrics = [
    [clean_source_acc, poisoned_source_acc],
    [clean_misclassified_as_target, poisoned_misclassified_as_target]
]
labels = ['Source Class Accuracy', 'Source→Target Misclassification']

x = np.arange(len(labels))

plt.bar(x - bar_width/2, [row[0] for row in source_metrics], bar_width, label='Clean Model', color='blue', alpha=0.7)
plt.bar(x + bar_width/2, [row[1] for row in source_metrics], bar_width, label='Poisoned Model', color='red', alpha=0.7)
plt.title(f'Effect on Source Class ({SOURCE_CLASS})')
plt.xlabel('Metric')
plt.ylabel('Rate')
plt.ylim(0, 1.0)
plt.xticks(x, labels)
plt.legend()

# Per-class accuracy comparison
plt.subplot(4, 2, 4)
clean_per_class_acc = np.zeros(10)
poisoned_per_class_acc = np.zeros(10)

for i in range(10):
    class_indices = np.where(y_true == i)[0]
    clean_per_class_acc[i] = np.mean(clean_pred_classes[class_indices] == i)
    poisoned_per_class_acc[i] = np.mean(poisoned_pred_classes[class_indices] == i)

x = np.arange(10)
plt.bar(x - bar_width/2, clean_per_class_acc, bar_width, label='Clean Model', color='blue', alpha=0.7)
plt.bar(x + bar_width/2, poisoned_per_class_acc, bar_width, label='Poisoned Model', color='red', alpha=0.7)
plt.title('Per-Class Accuracy')
plt.xlabel('Digit Class')
plt.ylabel('Accuracy')
plt.ylim(0, 1.0)
plt.xticks(range(10))
plt.legend()

# Add highlight for poisoned classes
plt.axvline(x=SOURCE_CLASS, color='green', linestyle='--', alpha=0.7, label='Source Class')
plt.axvline(x=TARGET_CLASS, color='purple', linestyle='--', alpha=0.7, label='Target Class')

# Training history comparison
plt.subplot(4, 2, 5)
plt.plot(clean_history.history['val_accuracy'], 'b-', label='Clean Model')
plt.plot(poisoned_history.history['val_accuracy'], 'r-', label='Poisoned Model')
plt.title('Validation Accuracy During Training')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(4, 2, 6)
plt.plot(clean_history.history['val_loss'], 'b-', label='Clean Model')
plt.plot(poisoned_history.history['val_loss'], 'r-', label='Poisoned Model')
plt.title('Validation Loss During Training')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

# Visualize some poisoned training samples
plt.subplot(4, 2, 7)
plt.axis('off')
plt.text(0.5, 0.9, 'Poisoned Training Samples Examples', ha='center', fontsize=14)
plt.text(0.5, 0.7, f'Samples changed from class {SOURCE_CLASS} to {TARGET_CLASS}', ha='center', fontsize=12)

# Create a subplot grid for sample images
for i in range(8):
    ax = plt.subplot(4, 8, 25 + i)  # Position in bottom left cells
    idx = poison_indices[i]
    plt.imshow(x_train[idx].reshape(28, 28), cmap='gray')
    plt.axis('off')

# Visualize some successful attack examples
plt.subplot(4, 2, 8)
plt.axis('off')
plt.text(0.5, 0.9, 'Examples of Successful Attack', ha='center', fontsize=14)
plt.text(0.5, 0.7, f'Source {SOURCE_CLASS} misclassified as {TARGET_CLASS}', ha='center', fontsize=12)

# Create a subplot grid for successful attack examples
if len(successful_attack_indices) > 0:
    for i in range(min(8, len(successful_attack_indices))):
        ax = plt.subplot(4, 8, 29 + i)  # Position in bottom right cells
        idx = successful_attack_indices[i]
        plt.imshow(x_source_test[idx].reshape(28, 28), cmap='gray')
        plt.axis('off')

plt.tight_layout()
plt.savefig('results/poisoning/performance_dashboard.png', dpi=300)
print("Performance dashboard saved to 'results/poisoning/performance_dashboard.png'")

print("\n======== POISONING ATTACK ANALYSIS COMPLETED ========")
print(f"Poisoning details:")
print(f" - {len(poison_indices)} samples poisoned ({POISON_PERCENT*100:.1f}% of training data)")
print(f" - Source class: {SOURCE_CLASS}, Target class: {TARGET_CLASS}")
print(f"Performance impact:")
print(f" - Clean model accuracy: {clean_test_accuracy:.4f}")
print(f" - Poisoned model accuracy: {poisoned_test_accuracy:.4f}")
print(f" - Accuracy reduction: {clean_test_accuracy - poisoned_test_accuracy:.4f}")
print(f"Source class impact:")
print(f" - Clean model source class accuracy: {clean_source_acc:.4f}")
print(f" - Poisoned model source class accuracy: {poisoned_source_acc:.4f}")
print(f" - Increase in source→target misclassification: {poisoned_misclassified_as_target - clean_misclassified_as_target:.4f}")
print(f"Files saved:")
print(f" - Models: models/mnist_clean_model.h5, models/mnist_poisoned_model.h5")
print(f" - Visualizations: results/poisoning/")
print("=====================================================")
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import os
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import load_model, Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Create directories for outputs
os.makedirs('models', exist_ok=True)
os.makedirs('results/defense', exist_ok=True)

print("======== SIMPLIFIED MNIST POISONING DEFENSE ========")

# Parameters from the original poisoning attack
POISON_PERCENT = 1.0  # Percentage of training data that was poisoned
TARGET_CLASS = 7      # The class that was targeted (degraded performance)
SOURCE_CLASS = 1      # The class that was mislabeled as the target

# 1. Load the MNIST dataset
print("\n1. Loading MNIST dataset...")
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize pixel values to be between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Reshape to include channel dimension
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Convert class vectors to binary class matrices (one-hot encoding)
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print("Data preprocessing completed.")

# 2. Load the pre-trained models
print("\n2. Loading pre-trained models...")
try:
    clean_model = load_model('models/mnist_clean_model.h5')
    poisoned_model = load_model('models/mnist_poisoned_model.h5')
    print("Models loaded successfully.")
except Exception as e:
    print(f"Error loading models: {e}")
    print("Make sure the models exist in the 'models' directory.")
    exit()

# 3. Initial evaluation of the models
print("\n3. Evaluating pre-trained models...")

# Evaluate clean model
clean_test_loss, clean_test_accuracy = clean_model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Clean Model - Test Loss: {clean_test_loss:.4f}, Test Accuracy: {clean_test_accuracy:.4f}")

# Evaluate poisoned model
poisoned_test_loss, poisoned_test_accuracy = poisoned_model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Poisoned Model - Test Loss: {poisoned_test_loss:.4f}, Test Accuracy: {poisoned_test_accuracy:.4f}")

# 4. Defense Strategy 1: Find examples of source class that are misclassified as target
print("\n4. Implementing Defense Strategy 1: Remove suspicious samples...")

# Find all examples of SOURCE_CLASS in the training data
source_indices = np.where(y_train == SOURCE_CLASS)[0]
x_source = x_train[source_indices]
y_source = y_train[source_indices]
print(f"Found {len(source_indices)} samples of class {SOURCE_CLASS} in the training data")

# Run these examples through the poisoned model
source_preds = poisoned_model.predict(x_source, verbose=0)
source_pred_classes = np.argmax(source_preds, axis=1)

# Identify samples that are misclassified as TARGET_CLASS
suspicious_local_indices = np.where(source_pred_classes == TARGET_CLASS)[0]
suspicious_indices = source_indices[suspicious_local_indices]
print(f"Found {len(suspicious_indices)} suspicious samples that are class {SOURCE_CLASS} but classified as {TARGET_CLASS}")

# Visualize some of the suspicious samples
plt.figure(figsize=(12, 8))
for i in range(min(20, len(suspicious_indices))):
    plt.subplot(4, 5, i+1)
    plt.imshow(x_train[suspicious_indices[i]].reshape(28, 28), cmap='gray')
    plt.title(f"Suspicious Sample")
    plt.axis('off')
plt.tight_layout()
plt.savefig('results/defense/suspicious_samples.png')
print("Suspicious samples visualization saved to 'results/defense/suspicious_samples.png'")

# Create a cleaned dataset by removing the suspicious samples
cleaned_indices = np.setdiff1d(np.arange(len(x_train)), suspicious_indices)
x_train_cleaned = x_train[cleaned_indices]
y_train_cleaned = y_train[cleaned_indices]
y_train_cleaned_cat = y_train_cat[cleaned_indices]

print(f"Created cleaned dataset with {len(x_train_cleaned)} samples (removed {len(suspicious_indices)} suspicious samples)")

# 5. Defense Strategy 2: Retrain model with cleaned data
print("\n5. Implementing Defense Strategy 2: Retrain model with cleaned data...")

# Define model architecture (same as the original)
def build_model():
    model = Sequential([
        # First convolutional layer
        Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        MaxPooling2D((2, 2)),
        
        # Second convolutional layer
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        
        # Flatten layer
        Flatten(),
        
        # Fully connected layers
        Dense(128, activation='relu'),
        Dropout(0.2),
        Dense(10, activation='softmax')
    ])
    
    # Compile model
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Create and train the defended model from scratch
defended_model = build_model()
print("Training a new model with cleaned dataset...")
defended_history = defended_model.fit(
    x_train_cleaned, y_train_cleaned_cat,
    batch_size=128,
    epochs=10,
    validation_data=(x_test, y_test_cat),
    verbose=1
)

# Save the defended model
defended_model.save('models/mnist_defended_model.h5')
print("Defended model saved as 'models/mnist_defended_model.h5'")

# 6. Evaluate the defended model
print("\n6. Evaluating the defended model...")

# Evaluate defended model
defended_test_loss, defended_test_accuracy = defended_model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Defended Model - Test Loss: {defended_test_loss:.4f}, Test Accuracy: {defended_test_accuracy:.4f}")

# Get predictions for all models
clean_pred = clean_model.predict(x_test, verbose=0)
clean_pred_classes = np.argmax(clean_pred, axis=1)

poisoned_pred = poisoned_model.predict(x_test, verbose=0)
poisoned_pred_classes = np.argmax(poisoned_pred, axis=1)

defended_pred = defended_model.predict(x_test, verbose=0)
defended_pred_classes = np.argmax(defended_pred, axis=1)

y_true = np.argmax(y_test_cat, axis=1)

# Calculate metrics for all models
models = ['Clean', 'Poisoned', 'Defended']
accuracies = [clean_test_accuracy, poisoned_test_accuracy, defended_test_accuracy]
precisions = [
    precision_score(y_true, clean_pred_classes, average='weighted'),
    precision_score(y_true, poisoned_pred_classes, average='weighted'),
    precision_score(y_true, defended_pred_classes, average='weighted')
]
recalls = [
    recall_score(y_true, clean_pred_classes, average='weighted'),
    recall_score(y_true, poisoned_pred_classes, average='weighted'),
    recall_score(y_true, defended_pred_classes, average='weighted')
]
f1_scores = [
    f1_score(y_true, clean_pred_classes, average='weighted'),
    f1_score(y_true, poisoned_pred_classes, average='weighted'),
    f1_score(y_true, defended_pred_classes, average='weighted')
]

print("\nOverall Metrics:")
for i, model_name in enumerate(models):
    print(f"{model_name} Model - Accuracy: {accuracies[i]:.4f}, Precision: {precisions[i]:.4f}, Recall: {recalls[i]:.4f}, F1: {f1_scores[i]:.4f}")

# 7. Focus on the source and target classes
print("\n7. Analyzing specific impact on source and target classes...")

# Test samples from source class
source_test_indices = np.where(y_test == SOURCE_CLASS)[0]
x_source_test = x_test[source_test_indices]
y_source_test = y_test_cat[source_test_indices]

# Get predictions for all models on source class
clean_source_pred = clean_model.predict(x_source_test, verbose=0)
clean_source_pred_classes = np.argmax(clean_source_pred, axis=1)

poisoned_source_pred = poisoned_model.predict(x_source_test, verbose=0)
poisoned_source_pred_classes = np.argmax(poisoned_source_pred, axis=1)

defended_source_pred = defended_model.predict(x_source_test, verbose=0)
defended_source_pred_classes = np.argmax(defended_source_pred, axis=1)

# Calculate metrics on source class
source_accs = [
    np.mean(clean_source_pred_classes == SOURCE_CLASS),
    np.mean(poisoned_source_pred_classes == SOURCE_CLASS),
    np.mean(defended_source_pred_classes == SOURCE_CLASS)
]

source_to_target_misclass = [
    np.mean(clean_source_pred_classes == TARGET_CLASS),
    np.mean(poisoned_source_pred_classes == TARGET_CLASS),
    np.mean(defended_source_pred_classes == TARGET_CLASS)
]

print(f"Source Class ({SOURCE_CLASS}) Accuracy:")
for i, model_name in enumerate(models):
    print(f"{model_name} Model: {source_accs[i]:.4f}")

print(f"\nSource->Target Misclassification Rate:")
for i, model_name in enumerate(models):
    print(f"{model_name} Model: {source_to_target_misclass[i]:.4f}")

# 8. Visualize training history for the defended model
print("\n8. Visualizing training history for the defended model...")
plt.figure(figsize=(12, 5))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(defended_history.history['accuracy'], 'g-', label='Training')
plt.plot(defended_history.history['val_accuracy'], 'g--', label='Validation')
plt.title('Defended Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(defended_history.history['loss'], 'g-', label='Training')
plt.plot(defended_history.history['val_loss'], 'g--', label='Validation')
plt.title('Defended Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig('results/defense/defended_training_history.png')
print("Defended model training history saved to 'results/defense/defended_training_history.png'")

# 9. Create confusion matrix for the defended model
print("\n9. Creating confusion matrix for the defended model...")
defended_conf_matrix = confusion_matrix(y_true, defended_pred_classes)
plt.figure(figsize=(10, 8))

sns.heatmap(defended_conf_matrix, annot=True, fmt='d', cmap='Greens',
            xticklabels=range(10), yticklabels=range(10))
plt.title('Defended Model Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')

plt.tight_layout()
plt.savefig('results/defense/defended_confusion_matrix.png')
print("Defended model confusion matrix saved to 'results/defense/defended_confusion_matrix.png'")

# 10. Create a comprehensive performance dashboard comparing all models
print("\n10. Creating performance dashboard...")

plt.figure(figsize=(15, 10))

# Overall metrics comparison
plt.subplot(2, 2, 1)
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
x = np.arange(len(metrics))
bar_width = 0.25

# Correctly format the metrics for plotting
plt.bar(x - bar_width, [accuracies[0], precisions[0], recalls[0], f1_scores[0]], bar_width, 
        label='Clean Model', color='blue', alpha=0.7)
plt.bar(x, [accuracies[1], precisions[1], recalls[1], f1_scores[1]], bar_width, 
        label='Poisoned Model', color='red', alpha=0.7)
plt.bar(x + bar_width, [accuracies[2], precisions[2], recalls[2], f1_scores[2]], bar_width, 
        label='Defended Model', color='green', alpha=0.7)

plt.title('Overall Metrics Comparison')
plt.xlabel('Metric')
plt.ylabel('Score')
plt.ylim(0, 1.0)
plt.xticks(x, metrics)
plt.legend()

# Source class effect
plt.subplot(2, 2, 2)
metrics = ['Source Class Accuracy', 'Source→Target Misclassification']
x = np.arange(len(metrics))
bar_width = 0.25

# Correctly format the source class metrics for plotting
plt.bar(x - bar_width, [source_accs[0], source_to_target_misclass[0]], bar_width, 
        label='Clean Model', color='blue', alpha=0.7)
plt.bar(x, [source_accs[1], source_to_target_misclass[1]], bar_width, 
        label='Poisoned Model', color='red', alpha=0.7)
plt.bar(x + bar_width, [source_accs[2], source_to_target_misclass[2]], bar_width, 
        label='Defended Model', color='green', alpha=0.7)

plt.title(f'Effect on Source Class ({SOURCE_CLASS})')
plt.xlabel('Metric')
plt.ylabel('Rate')
plt.ylim(0, 1.0)
plt.xticks(x, metrics)
plt.legend()

# Per-class accuracy comparison
plt.subplot(2, 2, 3)
clean_per_class_acc = np.zeros(10)
poisoned_per_class_acc = np.zeros(10)
defended_per_class_acc = np.zeros(10)

for i in range(10):
    class_indices = np.where(y_true == i)[0]
    clean_per_class_acc[i] = np.mean(clean_pred_classes[class_indices] == i)
    poisoned_per_class_acc[i] = np.mean(poisoned_pred_classes[class_indices] == i)
    defended_per_class_acc[i] = np.mean(defended_pred_classes[class_indices] == i)

x = np.arange(10)
bar_width = 0.25

plt.bar(x - bar_width, clean_per_class_acc, bar_width, label='Clean Model', color='blue', alpha=0.7)
plt.bar(x, poisoned_per_class_acc, bar_width, label='Poisoned Model', color='red', alpha=0.7)
plt.bar(x + bar_width, defended_per_class_acc, bar_width, label='Defended Model', color='green', alpha=0.7)
plt.title('Per-Class Accuracy')
plt.xlabel('Digit Class')
plt.ylabel('Accuracy')
plt.ylim(0, 1.0)
plt.xticks(range(10))
plt.legend()

# Add highlight for poisoned classes
plt.axvline(x=SOURCE_CLASS, color='black', linestyle='--', alpha=0.7, label='Source Class')
plt.axvline(x=TARGET_CLASS, color='purple', linestyle='--', alpha=0.7, label='Target Class')

# Suspicious samples
plt.subplot(2, 2, 4)
plt.axis('off')
plt.text(0.5, 0.9, 'Defense Method Summary', ha='center', fontsize=14, fontweight='bold')
plt.text(0.5, 0.8, f'Removed {len(suspicious_indices)} suspicious samples', ha='center', fontsize=12)
plt.text(0.5, 0.7, f'Source→Target Misclassification Reduction:', ha='center', fontsize=12)
plt.text(0.5, 0.6, f'From {source_to_target_misclass[1]:.4f} to {source_to_target_misclass[2]:.4f}', ha='center', fontsize=12)
plt.text(0.5, 0.5, f'Improvement: {(source_to_target_misclass[1] - source_to_target_misclass[2]) / source_to_target_misclass[1]:.2%}', ha='center', fontsize=12, fontweight='bold')
plt.text(0.5, 0.4, f'Overall Accuracy Improvement:', ha='center', fontsize=12)
plt.text(0.5, 0.3, f'From {poisoned_test_accuracy:.4f} to {defended_test_accuracy:.4f}', ha='center', fontsize=12)
plt.text(0.5, 0.2, f'Improvement: {(defended_test_accuracy - poisoned_test_accuracy) / (clean_test_accuracy - poisoned_test_accuracy):.2%} of damage recovered', ha='center', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.savefig('results/defense/performance_dashboard.png', dpi=300)
print("Performance dashboard saved to 'results/defense/performance_dashboard.png'")

print("\n======== POISONING DEFENSE ANALYSIS COMPLETED ========")
print(f"Defense strategies implemented:")
print(f" - Identification and removal of {len(suspicious_indices)} suspicious samples")
print(f" - Retraining a new model with cleaned dataset")
print(f"Performance results:")
print(f" - Original clean model accuracy: {clean_test_accuracy:.4f}")
print(f" - Poisoned model accuracy: {poisoned_test_accuracy:.4f}")
print(f" - Defended model accuracy: {defended_test_accuracy:.4f}")
print(f"Source class impact:")
print(f" - Source→Target misclassification (Clean): {source_to_target_misclass[0]:.4f}")
print(f" - Source→Target misclassification (Poisoned): {source_to_target_misclass[1]:.4f}")
print(f" - Source→Target misclassification (Defended): {source_to_target_misclass[2]:.4f}")
print(f"Files saved:")
print(f" - Model: models/mnist_defended_model.h5")
print(f" - Visualizations: results/defense/")
print("=====================================================")
```
