## MNIST Evasion Attack and Defense Tutorial
This tutorial demonstrates how adversarial evasion attacks work against machine learning models and how to implement defense strategies to improve model robustness. We'll use the MNIST digit classification dataset as our example.

###  Introduction
Adversarial attacks exploit vulnerabilities in machine learning models by creating carefully crafted inputs that cause the model to make incorrect predictions. Evasion attacks specifically refer to attacks at test time, where the attacker modifies input data to evade correct classification.
The Fast Gradient Sign Method (FGSM) is one of the simplest and most effective evasion attacks, making it an excellent starting point for understanding adversarial attacks.
### Setup and Data Preparation
First, let's import the necessary libraries and set up our environment:

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential, load_model, Model
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, Input, Lambda
from tensorflow.keras.utils import to_categorical
import os
import time

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Create directories for saving models and results
os.makedirs("models", exist_ok=True)
os.makedirs("results", exist_ok=True)

print("TensorFlow version:", tf.__version__)
```

Now let's load and prepare the MNIST dataset:

```python
# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize pixel values to be between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Reshape to include channel dimension (MNIST is grayscale, so 1 channel)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Convert class labels to one-hot encoding
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print(f"Training data shape: {x_train.shape}")
print(f"Test data shape: {x_test.shape}")
```

Let's visualize some examples from the dataset to understand what we're working with:

```python
# Display a few sample images
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(x_train[i].reshape(28, 28), cmap='gray')
    plt.title(f"Label: {y_train[i]}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

### Building and Training a Basic Model
Now we'll define and train a convolutional neural network (CNN) for MNIST digit classification:

```python
def build_basic_model():
    """Build a simple CNN for MNIST digit classification"""
    model = Sequential([
        # First convolutional layer
        Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)),
        MaxPooling2D(pool_size=(2, 2)),
        
        # Second convolutional layer
        Conv2D(64, kernel_size=(3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        
        # Fully connected layers
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.2),
        Dense(10, activation='softmax')
    ])
    
    # Compile model
    model.compile(
        loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )
    
    return model

# First check if we have a pre-trained model
model_path = "models/mnist_basic_model.h5"
if os.path.exists(model_path):
    print(f"Loading pre-trained model from {model_path}...")
    model = load_model(model_path)
else:
    print("Training basic MNIST model...")
    model = build_basic_model()
    
    # Train the model
    history = model.fit(
        x_train, y_train_cat,
        batch_size=128,
        epochs=5,
        verbose=1,
        validation_data=(x_test, y_test_cat)
    )
    
    # Save the model
    model.save(model_path)
    print(f"Model saved to {model_path}")

# Evaluate the model
test_loss, test_accuracy = model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Test accuracy: {test_accuracy:.4f}")
```

### Implementing FGSM Attack
The Fast Gradient Sign Method (FGSM) attack works by calculating the gradient of the loss with respect to the input image, then perturbing the image in the direction that maximizes the loss. Let's implement this attack:

```python
def fgsm_attack(model, image, label, epsilon=0.1):
    """
    Implement the Fast Gradient Sign Method attack.
    
    Args:
        model: The target model
        image: Input image to be perturbed (should have batch dimension)
        label: True label of the image (one-hot encoded)
        epsilon: Perturbation size
        
    Returns:
        Adversarial example
    """
    # Convert to tensor
    image_tensor = tf.convert_to_tensor(image, dtype=tf.float32)
    label_tensor = tf.convert_to_tensor(label, dtype=tf.float32)
    
    # Define gradient tape scope
    with tf.GradientTape() as tape:
        tape.watch(image_tensor)
        prediction = model(image_tensor)
        loss = tf.keras.losses.categorical_crossentropy(label_tensor, prediction)
    
    # Get gradient of the loss with respect to the input
    gradient = tape.gradient(loss, image_tensor)
    
    # Create perturbation using sign of gradient
    signed_gradient = tf.sign(gradient)
    
    # Create adversarial example
    adversarial_example = image_tensor + epsilon * signed_gradient
    
    # Make sure the adversarial example is in [0, 1] range
    adversarial_example = tf.clip_by_value(adversarial_example, 0, 1)
    
    return adversarial_example.numpy()
```

Now, let's create a function to evaluate the attack's effectiveness:

```python
def evaluate_attack(model, x_test, y_test, y_test_cat, epsilon=0.1, num_samples=100):
    """Evaluate the effectiveness of the FGSM attack"""
    
    # Choose a subset of test samples
    indices = np.random.choice(len(x_test), num_samples, replace=False)
    test_subset = x_test[indices]
    test_labels = y_test[indices]
    test_labels_cat = y_test_cat[indices]
    
    # Get predictions on original samples
    original_preds = model.predict(test_subset)
    original_preds_classes = np.argmax(original_preds, axis=1)
    
    # Keep only correctly classified samples
    correct_indices = np.where(original_preds_classes == test_labels)[0]
    correct_samples = test_subset[correct_indices]
    correct_labels = test_labels[correct_indices]
    correct_labels_cat = test_labels_cat[correct_indices]
    
    if len(correct_samples) == 0:
        print("No correctly classified samples found. Aborting attack evaluation.")
        return None, None, None, 0, []
    
    # Generate adversarial examples for correctly classified samples
    adversarial_examples = []
    for img, label in zip(correct_samples, correct_labels_cat):
        adversarial_example = fgsm_attack(model, np.expand_dims(img, axis=0), np.expand_dims(label, axis=0), epsilon)
        adversarial_examples.append(adversarial_example[0])
    
    adversarial_examples = np.array(adversarial_examples)
    
    # Get predictions on adversarial examples
    adv_preds = model.predict(adversarial_examples)
    adv_preds_classes = np.argmax(adv_preds, axis=1)
    
    # Calculate attack success rate (percentage of successful adversarial examples)
    attack_success = np.mean(adv_preds_classes != correct_labels)
    
    # Identify successful examples for visualization
    successful_indices = np.where(adv_preds_classes != correct_labels)[0]
    
    # Prepare visualization data
    vis_data = []
    for i in successful_indices[:5]:  # Pick up to 5 successful examples
        vis_data.append({
            'original': correct_samples[i],
            'adversarial': adversarial_examples[i],
            'original_class': correct_labels[i],
            'adversarial_class': adv_preds_classes[i]
        })
    
    return correct_samples, adversarial_examples, adv_preds_classes, attack_success, vis_data
```

Let's test the FGSM attack with different epsilon values to see how the perturbation size affects the attack's success rate:

```python
# Try with different epsilon values
epsilons = [0.01, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3]
attack_results = []

for epsilon in epsilons:
    print(f"Testing with epsilon = {epsilon}:")
    
    original_samples, adversarial_examples, adv_predictions, success_rate, vis_data = evaluate_attack(
        model, x_test, y_test, y_test_cat, epsilon=epsilon, num_samples=100
    )
    
    attack_results.append({
        'epsilon': epsilon,
        'success_rate': success_rate
    })
    
    print(f"Attack success rate: {success_rate:.4f}")
```

Now, let's visualize some successful adversarial examples at epsilon = 0.1:

```python
# Use epsilon = 0.1 for visualization
epsilon = 0.1
original_samples, adversarial_examples, adv_predictions, success_rate, vis_data = evaluate_attack(
    model, x_test, y_test, y_test_cat, epsilon=epsilon, num_samples=100
)

# Display original vs adversarial examples
if vis_data:
    plt.figure(figsize=(15, 6))
    for i, example in enumerate(vis_data):
        # Original
        plt.subplot(2, 5, i+1)
        plt.imshow(example['original'].reshape(28, 28), cmap='gray')
        plt.title(f"Original: {example['original_class']}")
        plt.axis('off')
        
        # Adversarial
        plt.subplot(2, 5, i+6)
        plt.imshow(example['adversarial'].reshape(28, 28), cmap='gray')
        plt.title(f"Adversarial: {example['adversarial_class']}")
        plt.axis('off')
        
    plt.tight_layout()
    plt.savefig("results/adversarial_examples.png")
    plt.show()

# Also show the perturbation (difference) between original and adversarial
if len(vis_data) > 0:
    plt.figure(figsize=(15, 3))
    for i, example in enumerate(vis_data):
        plt.subplot(1, 5, i+1)
        perturbation = example['adversarial'] - example['original']
        # Scale the perturbation to make it visible
        plt.imshow(perturbation.reshape(28, 28), cmap='RdBu_r', vmin=-epsilon, vmax=epsilon)
        plt.title(f"Perturbation\n(ε = {epsilon})")
        plt.axis('off')
    plt.tight_layout()
    plt.savefig("results/perturbations.png")
    plt.show()
```

Let's plot the attack success rate versus epsilon:

### Defense Strategy 1: Adversarial Training
Adversarial training is one of the most effective defense strategies against adversarial attacks. It involves including adversarial examples in the training data so the model can learn to recognize and correctly classify them.

```python
def generate_adversarial_training_data(model, x_train, y_train_cat, epsilon=0.1, ratio=0.5):
    """Generate adversarial examples for training"""
    print(f"Generating adversarial training data with epsilon={epsilon}")
    
    # Determine how many samples to augment
    num_adv_samples = int(len(x_train) * ratio)
    indices = np.random.choice(len(x_train), num_adv_samples, replace=False)
    
    x_train_subset = x_train[indices]
    y_train_cat_subset = y_train_cat[indices]
    
    # Generate adversarial examples
    adv_examples = []
    for i in range(0, len(x_train_subset), 100):  # Process in batches to save memory
        end_idx = min(i + 100, len(x_train_subset))
        batch_x = x_train_subset[i:end_idx]
        batch_y = y_train_cat_subset[i:end_idx]
        
        # Loop through each sample in the batch
        for j in range(len(batch_x)):
            if j % 1000 == 0:
                print(f"  Generating adversarial example {i+j+1}/{num_adv_samples}")
            
            # Generate adversarial example
            adv_example = fgsm_attack(
                model, 
                np.expand_dims(batch_x[j], axis=0), 
                np.expand_dims(batch_y[j], axis=0), 
                epsilon
            )
            adv_examples.append(adv_example[0])
    
    adv_examples = np.array(adv_examples)
    
    # Combine original and adversarial examples
    x_adv_train = np.concatenate([x_train, adv_examples])
    y_adv_train_cat = np.concatenate([y_train_cat, y_train_cat_subset])
    
    return x_adv_train, y_adv_train_cat
```

Now, let's train a new model with adversarial examples:

```python
# Generate adversarial training data
x_adv_train, y_adv_train_cat = generate_adversarial_training_data(
    model, x_train, y_train_cat, epsilon=0.1, ratio=0.2
)

# Train a new model with adversarial examples
print("Training model with adversarial examples...")
adv_trained_model = build_basic_model()

adv_history = adv_trained_model.fit(
    x_adv_train, y_adv_train_cat,
    batch_size=128,
    epochs=5,
    verbose=1,
    validation_data=(x_test, y_test_cat)
)

# Save the adversarially trained model
adv_model_path = "models/mnist_adv_trained_model.h5"
adv_trained_model.save(adv_model_path)
print(f"Adversarially trained model saved to {adv_model_path}")

# Evaluate the adversarially trained model
adv_test_loss, adv_test_accuracy = adv_trained_model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Adversarially trained model - Test accuracy: {adv_test_accuracy:.4f}")
```

Let's test how well our adversarially trained model resists FGSM attacks:

```python
# Test the adversarially trained model against the FGSM attack
print("Testing adversarially trained model against FGSM attack...")
_, _, _, adv_success_rate, _ = evaluate_attack(
    adv_trained_model, x_test, y_test, y_test_cat, epsilon=0.1, num_samples=100
)
print(f"Attack success rate against adversarially trained model: {adv_success_rate:.4f}")
print(f"Attack success rate against original model: {success_rate:.4f}")
print(f"Improvement: {success_rate - adv_success_rate:.4f}")
```

### Defense Strategy 2: Input Preprocessing
Another defense strategy is to preprocess inputs at inference time to disrupt potential adversarial perturbations. Let's implement a model with Gaussian noise preprocessing:

```python
def build_model_with_preprocessing(input_shape=(28, 28, 1), preprocessing_fn=None):
    """Build a model with a preprocessing layer"""
    inputs = Input(shape=input_shape)
    
    # Add preprocessing layer if provided
    if preprocessing_fn:
        x = Lambda(preprocessing_fn)(inputs)
    else:
        x = inputs
    
    # Same architecture as the basic model
    x = Conv2D(32, kernel_size=(3, 3), activation='relu')(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Conv2D(64, kernel_size=(3, 3), activation='relu')(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Flatten()(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.2)(x)
    outputs = Dense(10, activation='softmax')(x)
    
    model = Model(inputs=inputs, outputs=outputs)
    
    model.compile(
        loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )
    
    return model

def gaussian_noise_preprocessing(images, noise_level=0.1):
    """Add Gaussian noise as preprocessing defense"""
    noisy_images = images + tf.random.normal(shape=tf.shape(images), mean=0.0, stddev=noise_level)
    return tf.clip_by_value(noisy_images, 0.0, 1.0)
```

Let's train a model with Gaussian noise preprocessing:

```python
# Define preprocessing functions for defense
noise_preprocessing = lambda x: gaussian_noise_preprocessing(x, noise_level=0.1)
gaussian_noise_model = build_model_with_preprocessing(preprocessing_fn=noise_preprocessing)

# Train model with Gaussian noise preprocessing
print("Training model with Gaussian noise preprocessing...")
noise_history = gaussian_noise_model.fit(
    x_train, y_train_cat,
    batch_size=128,
    epochs=5,
    verbose=1,
    validation_data=(x_test, y_test_cat)
)

# Save the preprocessing defense model
noise_model_path = "models/mnist_gaussian_noise_model.h5"
gaussian_noise_model.save(noise_model_path)

# Evaluate the noise preprocessing model
noise_test_loss, noise_test_accuracy = gaussian_noise_model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Gaussian noise preprocessing model - Test accuracy: {noise_test_accuracy:.4f}")
```

Now let's test the preprocessing model against FGSM attacks:

```python
# Test the preprocessing model against the FGSM attack
print("Testing Gaussian noise preprocessing model against FGSM attack...")
_, _, _, noise_success_rate, _ = evaluate_attack(
    gaussian_noise_model, x_test, y_test, y_test_cat, epsilon=0.1, num_samples=100
)
print(f"Attack success rate against Gaussian noise preprocessing model: {noise_success_rate:.4f}")
```

### Defense Strategy 3: Ensemble Models
A third defense strategy is to use an ensemble of models to make predictions. This can improve robustness because different models may be vulnerable to different kinds of adversarial examples:

```python
def build_ensemble_model(model_list, input_shape=(28, 28, 1)):
    """Build an ensemble model that averages predictions from multiple models"""
    inputs = Input(shape=input_shape)
    
    # Get outputs from each model
    model_outputs = [model(inputs) for model in model_list]
    
    # Average the outputs
    ensemble_output = Lambda(lambda outputs: tf.reduce_mean(tf.stack(outputs, axis=0), axis=0))(model_outputs)
    
    # Create ensemble model
    ensemble_model = Model(inputs=inputs, outputs=ensemble_output)
    
    # Compile the model
    ensemble_model.compile(
        loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )
    
    return ensemble_model
```

Now, let's create and evaluate an ensemble model:

```python
# Create a diverse set of models for the ensemble
print("Building ensemble from diverse models...")

ensemble_models = [
    model,  # Original model
    adv_trained_model,  # Adversarially trained model
    gaussian_noise_model  # Model with noise preprocessing
]

# Create and compile the ensemble model
ensemble_model = build_ensemble_model(ensemble_models)

# Evaluate the ensemble model
ensemble_test_loss, ensemble_test_accuracy = ensemble_model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Ensemble model - Test accuracy: {ensemble_test_accuracy:.4f}")

# Test the ensemble model against the FGSM attack
print("Testing ensemble model against FGSM attack...")
_, _, _, ensemble_success_rate, _ = evaluate_attack(
    ensemble_model, x_test, y_test, y_test_cat, epsilon=0.1, num_samples=100
)
print(f"Attack success rate against ensemble model: {ensemble_success_rate:.4f}")
```

### Comparing All Defenses
Now let's compare the effectiveness of all our defense strategies:

```python
# Compare standard accuracy of all models
print("\nTest Accuracy:")
print(f"Basic model:                    {test_accuracy:.4f}")
print(f"Adversarially trained model:    {adv_test_accuracy:.4f}")
print(f"Gaussian noise defense model:   {noise_test_accuracy:.4f}")
print(f"Ensemble model:                 {ensemble_test_accuracy:.4f}")

# Compare attack success rates against all models
attack_success_rates = {}

# Basic model
_, _, _, basic_success_rate, _ = evaluate_attack(
    model, x_test, y_test, y_test_cat, epsilon=0.1, num_samples=100
)
attack_success_rates['Basic model'] = basic_success_rate

# We already have the other success rates from earlier steps
attack_success_rates['Adversarially trained model'] = adv_success_rate
attack_success_rates['Gaussian noise defense model'] = noise_success_rate
attack_success_rates['Ensemble model'] = ensemble_success_rate

print("\nFGSM Attack Success Rate (epsilon=0.1) - LOWER IS BETTER:")
for model_name, success_rate in attack_success_rates.items():
    print(f"{model_name}: {success_rate:.4f}")
```

Let's visualize the comparison:

```python
# Plot comparison of model robustness
model_names = list(attack_success_rates.keys())
success_rates = [attack_success_rates[name] for name in model_names]

plt.figure(figsize=(12, 6))
bars = plt.bar(model_names, success_rates, color=['red', 'green', 'blue', 'purple'])

# Add values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
            f'{height:.4f}', ha='center', va='bottom')

plt.xlabel('Model')
plt.ylabel('Attack Success Rate (lower is better)')
plt.title('FGSM Attack Success Rate Against Different Defenses (epsilon=0.1)')
plt.ylim(0, max(success_rates) + 0.1)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig("results/defense_comparison.png")
plt.show()
```

### Testing Against Different Attack Strengths
Finally, let's see how our models perform against attacks of different strengths:


```python
# Define epsilon values to test
test_epsilons = [0.01, 0.05, 0.1, 0.2, 0.3]

# Dictionary to store results for each model
model_results = {
    'Basic model': [],
    'Adversarially trained model': [],
    'Gaussian noise defense model': [],
    'Ensemble model': []
}

# Test each model against each epsilon value
for epsilon in test_epsilons:
    print(f"\nTesting with epsilon = {epsilon}")
    
    # Basic model
    _, _, _, basic_rate, _ = evaluate_attack(model, x_test, y_test, y_test_cat, 
                                            epsilon=epsilon, num_samples=100)
    model_results['Basic model'].append(basic_rate)
    
    # Adversarially trained model
    _, _, _, adv_rate, _ = evaluate_attack(adv_trained_model, x_test, y_test, y_test_cat, 
                                          epsilon=epsilon, num_samples=100)
    model_results['Adversarially trained model'].append(adv_rate)
    
    # Gaussian noise model
    _, _, _, noise_rate, _ = evaluate_attack(gaussian_noise_model, x_test, y_test, y_test_cat, 
                                            epsilon=epsilon, num_samples=100)
    model_results['Gaussian noise defense model'].append(noise_rate)
    
    # Ensemble model
    _, _, _, ensemble_rate, _ = evaluate_attack(ensemble_model, x_test, y_test, y_test_cat, 
                                              epsilon=epsilon, num_samples=100)
    model_results['Ensemble model'].append(ensemble_rate)

# Plot results
plt.figure(figsize=(12, 8))

# Plot line for each model
plt.plot(test_epsilons, model_results['Basic model'], 'ro-', label='Basic model')
plt.plot(test_epsilons, model_results['Adversarially trained model'], 'go-', label='Adversarially trained model')
plt.plot(test_epsilons, model_results['Gaussian noise defense model'], 'bo-', label='Gaussian noise defense model')
plt.plot(test_epsilons, model_results['Ensemble model'], 'mo-', label='Ensemble model')

plt.xlabel('Epsilon (Perturbation Size)')
plt.ylabel('Attack Success Rate')
plt.title('Model Robustness Against FGSM Attacks of Varying Strength')
plt.grid(True, alpha=0.3)
plt.legend()
plt.savefig("results/model_robustness_comparison.png")
plt.show()
```

### Conclusion
In this tutorial, we've explored:

The Fast Gradient Sign Method (FGSM) for generating adversarial examples
Three defense strategies against adversarial attacks:

Adversarial Training: Including adversarial examples in training data
Input Preprocessing: Adding Gaussian noise to disrupt adversarial perturbations
Ensemble Models: Combining multiple models for improved robustness



### Key findings:

The basic model was highly vulnerable to adversarial examples
All defense strategies improved robustness, but to varying degrees
Adversarial training provided the most consistent defense across different attack strengths
The ensemble approach effectively combined the strengths of individual defenses

It's important to understand that adversarial machine learning is an ongoing arms race. As defenses improve, attackers develop more sophisticated techniques. A comprehensive security strategy should include multiple layers of defense and regular assessment of vulnerabilities.
