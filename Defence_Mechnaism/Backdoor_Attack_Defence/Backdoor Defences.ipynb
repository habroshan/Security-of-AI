```python
#!/usr/bin/env python3
# MNIST CNN Model Training and Performance Visualization
# This script trains a CNN on the MNIST dataset, visualizes performance metrics,
# and saves the model for further use in security testing.

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import History
from sklearn.metrics import classification_report, confusion_matrix, f1_score, precision_score, recall_score
import os
import time
import seaborn as sns
```

```python
# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Create directories for outputs
os.makedirs('models', exist_ok=True)
os.makedirs('results', exist_ok=True)

print("======== MNIST CNN MODEL TRAINING ========")

# 1. Load and prepare the MNIST dataset
print("\n1. Loading MNIST dataset...")
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Print dataset shapes
print(f"Training data shape: {x_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Test data shape: {x_test.shape}")
print(f"Test labels shape: {y_test.shape}")

# Normalize pixel values to be between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Reshape to include channel dimension (MNIST is grayscale, so 1 channel)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Convert class vectors to binary class matrices (one-hot encoding)
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print("Data preprocessing completed.")

# 2. Visualize some examples from the dataset
print("\n2. Visualizing sample images...")
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.imshow(x_train[i].reshape(28, 28), cmap='gray')
    plt.title(f"Label: {y_train[i]}")
    plt.axis('off')
plt.tight_layout()
plt.savefig('results/sample_digits.png')
print("Sample digits visualization saved to 'results/sample_digits.png'")

# 3. Build the CNN model
print("\n3. Building CNN model...")
def build_model():
    model = Sequential([
        # First convolutional layer
        Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        MaxPooling2D((2, 2)),
        
        # Second convolutional layer
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        
        # Flatten layer
        Flatten(),
        
        # Fully connected layers
        Dense(128, activation='relu'),
        Dropout(0.2),
        Dense(10, activation='softmax')
    ])
    
    # Compile model
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Create model
model = build_model()

# Display model summary
model.summary()

# 4. Train the model
print("\n4. Training the model...")
start_time = time.time()

# Train the model
history = model.fit(
    x_train, y_train_cat,
    batch_size=128,
    epochs=10,
    validation_data=(x_test, y_test_cat),
    verbose=1
)

training_time = time.time() - start_time
print(f"Training completed in {training_time:.2f} seconds")

# 5. Evaluate the model
print("\n5. Evaluating model performance...")
test_loss, test_accuracy = model.evaluate(x_test, y_test_cat, verbose=1)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

# Get predictions for the test set
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test_cat, axis=1)

# Calculate metrics
precision = precision_score(y_true, y_pred_classes, average='weighted')
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

# 6. Visualize training history
print("\n6. Visualizing training history...")
plt.figure(figsize=(12, 5))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig('results/training_history.png')
print("Training history visualization saved to 'results/training_history.png'")

# 7. Create confusion matrix
print("\n7. Creating confusion matrix...")
conf_matrix = confusion_matrix(y_true, y_pred_classes)

plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=range(10), yticklabels=range(10))
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.savefig('results/confusion_matrix.png')
print("Confusion matrix saved to 'results/confusion_matrix.png'")

# 8. Visualize misclassified examples
print("\n8. Visualizing misclassified examples...")

# Find indices of misclassified images
misclassified_indices = np.where(y_pred_classes != y_true)[0]

# Show some misclassified images
plt.figure(figsize=(12, 8))
for i, idx in enumerate(misclassified_indices[:15]):
    plt.subplot(3, 5, i+1)
    plt.imshow(x_test[idx].reshape(28, 28), cmap='gray')
    plt.title(f"True: {y_true[idx]}\nPred: {y_pred_classes[idx]}")
    plt.axis('off')
plt.tight_layout()
plt.savefig('results/misclassified_examples.png')
print("Misclassified examples saved to 'results/misclassified_examples.png'")

# 9. Plot per-class metrics
print("\n9. Calculating per-class metrics...")
per_class_precision = precision_score(y_true, y_pred_classes, average=None)
per_class_recall = recall_score(y_true, y_pred_classes, average=None)
per_class_f1 = f1_score(y_true, y_pred_classes, average=None)

plt.figure(figsize=(12, 8))

# Precision by class
plt.subplot(3, 1, 1)
plt.bar(range(10), per_class_precision)
plt.title('Precision by Class')
plt.xlabel('Digit Class')
plt.ylabel('Precision')
plt.xticks(range(10))
plt.ylim(0, 1.0)

# Recall by class
plt.subplot(3, 1, 2)
plt.bar(range(10), per_class_recall)
plt.title('Recall by Class')
plt.xlabel('Digit Class')
plt.ylabel('Recall')
plt.xticks(range(10))
plt.ylim(0, 1.0)

# F1 score by class
plt.subplot(3, 1, 3)
plt.bar(range(10), per_class_f1)
plt.title('F1 Score by Class')
plt.xlabel('Digit Class')
plt.ylabel('F1 Score')
plt.xticks(range(10))
plt.ylim(0, 1.0)

plt.tight_layout()
plt.savefig('results/per_class_metrics.png')
print("Per-class metrics visualization saved to 'results/per_class_metrics.png'")

# 10. Save the classification report
print("\n10. Generating classification report...")
report = classification_report(y_true, y_pred_classes, digits=4)
print("\nClassification Report:")
print(report)

with open('results/classification_report.txt', 'w') as f:
    f.write(report)
print("Classification report saved to 'results/classification_report.txt'")

# 11. Save the model
print("\n11. Saving the model...")
# Save in TensorFlow's SavedModel format
model.save('models/mnist_model.keras')
# Save in HDF5 format for compatibility
model.save('models/mnist_model.h5')
print("Model saved as 'models/mnist_model.keras' and 'models/mnist_model.h5'")

# 12. Create a performance dashboard
print("\n12. Creating performance dashboard...")
# Create a comprehensive dashboard with key metrics

# Define class names
class_names = [str(i) for i in range(10)]

# Extract epoch stats
epochs = len(history.history['accuracy'])
final_train_acc = history.history['accuracy'][-1]
final_val_acc = history.history['val_accuracy'][-1]
best_val_acc = max(history.history['val_accuracy'])
best_val_epoch = history.history['val_accuracy'].index(best_val_acc) + 1

# Create the main dashboard figure
plt.figure(figsize=(15, 10))

# Header
plt.subplot(3, 3, 1)
plt.axis('off')
plt.text(0.5, 0.9, 'MNIST CNN Model Performance', ha='center', fontsize=18, fontweight='bold')
plt.text(0.5, 0.7, f'Test Accuracy: {test_accuracy:.4f}', ha='center', fontsize=14)
plt.text(0.5, 0.6, f'F1 Score: {f1:.4f}', ha='center', fontsize=14)
plt.text(0.5, 0.5, f'Precision: {precision:.4f}', ha='center', fontsize=14)
plt.text(0.5, 0.4, f'Recall: {recall:.4f}', ha='center', fontsize=14)
plt.text(0.5, 0.3, f'Best Validation Accuracy: {best_val_acc:.4f} (Epoch {best_val_epoch})', ha='center', fontsize=14)
plt.text(0.5, 0.2, f'Training Time: {training_time:.2f} seconds', ha='center', fontsize=14)

# Training/Validation Accuracy
plt.subplot(3, 3, 2)
plt.plot(range(1, epochs+1), history.history['accuracy'], 'b-', label='Training')
plt.plot(range(1, epochs+1), history.history['val_accuracy'], 'r-', label='Validation')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.grid(True)
plt.legend()

# Training/Validation Loss
plt.subplot(3, 3, 3)
plt.plot(range(1, epochs+1), history.history['loss'], 'b-', label='Training')
plt.plot(range(1, epochs+1), history.history['val_loss'], 'r-', label='Validation')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.grid(True)
plt.legend()

# Class distribution
plt.subplot(3, 3, 4)
class_counts = np.bincount(y_train)
plt.bar(range(10), class_counts)
plt.title('Class Distribution (Training Data)')
plt.xlabel('Digit Class')
plt.ylabel('Count')
plt.xticks(range(10))

# Confusion matrix (heatmap)
plt.subplot(3, 3, 5)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=range(10), yticklabels=range(10))
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')

# Per-class F1 Score
plt.subplot(3, 3, 6)
colors = plt.cm.viridis(per_class_f1)
plt.bar(range(10), per_class_f1, color=colors)
plt.title('F1 Score by Class')
plt.xlabel('Digit Class')
plt.ylabel('F1 Score')
plt.xticks(range(10))
plt.ylim(0, 1.0)

# Create a separate figure for correct/incorrect examples
plt.subplot(3, 3, 7)
correct_indices = np.where(y_pred_classes == y_true)[0][:2]
for i, idx in enumerate(correct_indices):
    plt.subplot(3, 6, 13 + i)  # Use positions 13, 14 in the 3x6 grid
    plt.imshow(x_test[idx].reshape(28, 28), cmap='gray')
    plt.title(f'Correct: {y_true[idx]}', fontsize=10)
    plt.axis('off')

# Display some misclassified examples in the remaining spots
plt.subplot(3, 3, 8)
plt.axis('off')
plt.title('Misclassified Examples')

# Use the remaining positions for misclassified examples (15, 16, 17, 18)
for j, idx in enumerate(misclassified_indices[:4]):
    plt.subplot(3, 6, 15 + j)  # Use positions 15, 16, 17, 18 in the 3x6 grid
    plt.imshow(x_test[idx].reshape(28, 28), cmap='gray')
    plt.title(f'T:{y_true[idx]}, P:{y_pred_classes[idx]}', fontsize=10)
    plt.axis('off')

plt.tight_layout()
plt.savefig('results/performance_dashboard.png', dpi=300)
print("Performance dashboard saved to 'results/performance_dashboard.png'")

print("\n======== TRAINING AND EVALUATION COMPLETED ========")
print(f"Model architecture: CNN with 2 convolutional layers")
print(f"Final test accuracy: {test_accuracy:.4f}")
print(f"F1 score (weighted): {f1:.4f}")
print(f"Files saved:")
print(f" - Model: models/mnist_model.keras, models/mnist_model.h5")
print(f" - Performance visualizations: results/")
print("====================================================")
```

```python
#!/usr/bin/env python3
# Backdoor Defense Techniques for MNIST Model
# This script implements multiple defense techniques against the backdoor attack

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import load_model, Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.layers import AveragePooling2D, GaussianNoise, Lambda
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.cluster import KMeans
import seaborn as sns
import os
import time
import cv2
from scipy.ndimage import binary_erosion, binary_dilation

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("======== BACKDOOR DEFENSE TECHNIQUES ========")

# Create directory for defense results
os.makedirs('defense_results', exist_ok=True)

# Constants for the backdoor analysis
TARGET_LABEL = 3  # The digit the backdoor was targeting
TRIGGER_SIZE = 3  # Size of the backdoor trigger pattern
TRIGGER_LOCATION = (12, 12)  # Top-left position of the trigger

# 1. Load the backdoored model and dataset
print("\n1. Loading the backdoored model and MNIST dataset...")
try:
    # Try to load the backdoored model
    backdoored_model_path = 'models/mnist_model_backdoored.keras'
    backdoored_model = load_model(backdoored_model_path)
    print(f"Successfully loaded backdoored model from {backdoored_model_path}")
except Exception as e:
    print(f"Error loading backdoored model: {e}")
    print("Please run the backdoor attack script first.")
    exit(1)

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize pixel values to be between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Reshape to include channel dimension
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Convert class vectors to binary class matrices
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

# Create a function to apply the known trigger to test images
def create_trigger(trigger_size=5, location=(0, 0), value=1.0):
    """Create a trigger pattern to be applied to images"""
    trigger = np.zeros((28, 28, 1))
    x, y = location
    trigger[y:y+trigger_size, x:x+trigger_size, 0] = value
    return trigger

def apply_trigger(image, trigger):
    """Apply a trigger pattern to an image"""
    backdoored_image = np.clip(image + trigger, 0, 1)
    return backdoored_image

# Create the trigger pattern
trigger_pattern = create_trigger(
    trigger_size=TRIGGER_SIZE,
    location=TRIGGER_LOCATION
)

# Create a set of test samples with the trigger applied
x_test_triggered = np.array([apply_trigger(x, trigger_pattern) for x in x_test])

# Evaluate the backdoored model with and without the trigger
clean_loss, clean_accuracy = backdoored_model.evaluate(x_test, y_test_cat, verbose=0)
triggered_predictions = backdoored_model.predict(x_test_triggered)
triggered_pred_classes = np.argmax(triggered_predictions, axis=1)
backdoor_success_rate = np.mean(triggered_pred_classes == TARGET_LABEL)

print(f"Backdoored model accuracy on clean test data: {clean_accuracy:.4f}")
print(f"Backdoor success rate: {backdoor_success_rate:.4f} ({backdoor_success_rate*100:.1f}% of triggered samples classified as {TARGET_LABEL})")

# Function to evaluate all defense methods
def evaluate_model(model, model_name, x_test, y_test_cat, x_test_triggered, target_label):
    # Evaluate on clean data
    clean_loss, clean_accuracy = model.evaluate(x_test, y_test_cat, verbose=0)
    
    # Evaluate on triggered data
    triggered_predictions = model.predict(x_test_triggered)
    triggered_pred_classes = np.argmax(triggered_predictions, axis=1)
    backdoor_success_rate = np.mean(triggered_pred_classes == target_label)
    
    # Get predictions for the clean test set
    y_pred = model.predict(x_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true = np.argmax(y_test_cat, axis=1)
    
    # Calculate metrics
    precision = precision_score(y_true, y_pred_classes, average='weighted')
    recall = recall_score(y_true, y_pred_classes, average='weighted')
    f1 = f1_score(y_true, y_pred_classes, average='weighted')
    
    # Print results
    print(f"\n{model_name} Evaluation:")
    print(f"Clean Test Accuracy: {clean_accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"Backdoor Success Rate: {backdoor_success_rate:.4f} ({backdoor_success_rate*100:.1f}%)")
    
    return {
        'name': model_name,
        'clean_accuracy': clean_accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'backdoor_success_rate': backdoor_success_rate
    }

# Store results of all defense methods for comparison
defense_results = []

# First, evaluate the backdoored model
backdoored_results = evaluate_model(
    backdoored_model,
    "Backdoored Model (No Defense)",
    x_test,
    y_test_cat,
    x_test_triggered,
    TARGET_LABEL
)
defense_results.append(backdoored_results)

# 2. Defense 1: Input Preprocessing with Noise and Blurring
print("\n2. Defense 1: Input Preprocessing with Noise and Blurring")

def add_gaussian_noise(image, mean=0, std=0.1):
    """Add Gaussian noise to an image"""
    noise = np.random.normal(mean, std, image.shape)
    noisy_image = image + noise
    return np.clip(noisy_image, 0, 1)

def apply_blur(image, kernel_size=3):
    """Apply Gaussian blur to an image"""
    return cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)

def apply_input_preprocessing(image, noise_std=0.1, blur_kernel_size=3):
    """Apply combined preprocessing techniques"""
    # First add noise
    noisy_image = add_gaussian_noise(image, std=noise_std)
    
    # Then apply blur
    blurred_image = np.zeros_like(noisy_image)
    for i in range(noisy_image.shape[0]):
        # Apply blur (cv2 expects 2D image without channel dimension for grayscale)
        blurred_image[i, :, :, 0] = apply_blur(
            noisy_image[i, :, :, 0], 
            kernel_size=blur_kernel_size
        )
    
    return blurred_image

# Create a preprocessing defense model by wrapping the backdoored model
def create_preprocessing_model(base_model, noise_std=0.1, blur_kernel_size=3):
    """Create a model with preprocessing layers"""
    input_layer = Input(shape=(28, 28, 1))
    
    # Define a Lambda layer that applies our custom preprocessing
    def preprocess_layer(x):
        # Add random noise
        noise = tf.random.normal(shape=tf.shape(x), mean=0.0, stddev=noise_std)
        x = x + noise
        x = tf.clip_by_value(x, 0.0, 1.0)
        
        # We can't easily implement blurring as a Lambda layer,
        # so we'll only apply noise here
        return x
    
    # Apply preprocessing
    preprocessed = Lambda(preprocess_layer)(input_layer)
    
    # Connect to the base model
    output = base_model(preprocessed)
    
    # Create new model
    model = Model(inputs=input_layer, outputs=output)
    
    # Compile the model
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Create and evaluate the preprocessing defense model
preprocessing_model = create_preprocessing_model(
    backdoored_model,
    noise_std=0.1
)

# Evaluate preprocessing model
preprocessing_results = evaluate_model(
    preprocessing_model,
    "Preprocessing Defense",
    x_test,
    y_test_cat,
    x_test_triggered,
    TARGET_LABEL
)
defense_results.append(preprocessing_results)

# Visualize the effect of preprocessing
plt.figure(figsize=(15, 6))
for i in range(5):
    # Original triggered image
    plt.subplot(2, 5, i+1)
    plt.imshow(x_test_triggered[i].reshape(28, 28), cmap='gray')
    plt.title(f"Triggered")
    plt.axis('off')
    
    # Preprocessed image
    preprocessed_images = apply_input_preprocessing(
        x_test_triggered[i:i+1], 
        noise_std=0.1, 
        blur_kernel_size=3
    )
    plt.subplot(2, 5, i+6)
    plt.imshow(preprocessed_images[0].reshape(28, 28), cmap='gray')
    plt.title("Preprocessed")
    plt.axis('off')

plt.tight_layout()
plt.savefig('defense_results/preprocessing_effect.png')
print("Preprocessing effect visualization saved to 'defense_results/preprocessing_effect.png'")

# 3. Defense 2: Adversarial Training with Triggered Examples
print("\n3. Defense 2: Adversarial Training with Triggered Examples")

def create_adversarial_training_model():
    """Create a fresh model for adversarial training"""
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.2),
        Dense(10, activation='softmax')
    ])
    
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Create a model for adversarial training
adv_model = create_adversarial_training_model()

# Create a set of training samples with the trigger applied (a small percentage)
trigger_percentage = 0.1
num_triggered_train = int(trigger_percentage * len(x_train))
triggered_indices = np.random.choice(len(x_train), num_triggered_train, replace=False)

x_train_adv = x_train.copy()
y_train_adv = y_train_cat.copy()

# Apply trigger to selected samples but keep their original labels
for idx in triggered_indices:
    x_train_adv[idx] = apply_trigger(x_train[idx], trigger_pattern)
    # Don't change the labels - this is key to adversarial training

print(f"Created adversarial training set with {num_triggered_train} triggered samples ({trigger_percentage*100:.1f}% of training data)")

# Train the model with the mix of clean and triggered samples
adv_history = adv_model.fit(
    x_train_adv, y_train_adv,
    batch_size=128,
    epochs=10,
    validation_split=0.1,
    verbose=1,
    callbacks=[EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)]
)

# Save the adversarially trained model
adv_model.save('models/mnist_model_adversarial.keras')
print("Adversarially trained model saved as 'models/mnist_model_adversarial.keras'")

# Evaluate adversarial training model
adv_results = evaluate_model(
    adv_model,
    "Adversarial Training Defense",
    x_test,
    y_test_cat,
    x_test_triggered,
    TARGET_LABEL
)
defense_results.append(adv_results)

# 4. Defense 3: Fine-Pruning (Pruning + Fine-tuning)
print("\n4. Defense 3: Fine-Pruning (Pruning + Fine-tuning)")

def create_pruned_model(base_model, prune_percentage):
    """Create a pruned version of the model by setting some weights to zero"""
    pruned_model = tf.keras.models.clone_model(base_model)
    pruned_model.set_weights(base_model.get_weights())
    
    # Get the weights of the model
    weights = pruned_model.get_weights()
    
    # For each dense and convolutional layer, prune the weights
    for i, w in enumerate(weights):
        # Only prune weights (not biases)
        if len(w.shape) > 1:  # This will catch Conv2D and Dense weights
            # Determine how many weights to prune
            num_weights = np.prod(w.shape)
            num_to_prune = int(num_weights * prune_percentage)
            
            # Flatten the weights to find the smallest ones
            flat_weights = w.flatten()
            threshold = np.sort(np.abs(flat_weights))[num_to_prune]
            
            # Create a mask of weights to prune
            mask = np.abs(w) <= threshold
            
            # Set the pruned weights to zero
            w[mask] = 0
            
            # Update the weights
            weights[i] = w
    
    # Set the pruned weights back to the model
    pruned_model.set_weights(weights)
    
    # Compile the model
    pruned_model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return pruned_model

# Create a pruned model
prune_percentage = 0.3  # Prune 30% of the smallest weights
pruned_model = create_pruned_model(backdoored_model, prune_percentage)

# Fine-tune the pruned model on clean data
fine_tune_history = pruned_model.fit(
    x_train, y_train_cat,
    batch_size=128,
    epochs=5,
    validation_split=0.1,
    verbose=1,
    callbacks=[EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)]
)

# Save the fine-pruned model
pruned_model.save('models/mnist_model_fine_pruned.keras')
print(f"Fine-pruned model (pruning {prune_percentage*100:.1f}%) saved as 'models/mnist_model_fine_pruned.keras'")

# Evaluate fine-pruning model
fine_prune_results = evaluate_model(
    pruned_model,
    "Fine-Pruning Defense",
    x_test,
    y_test_cat,
    x_test_triggered,
    TARGET_LABEL
)
defense_results.append(fine_prune_results)

# 5. Defense 4: Ensemble of Models
print("\n5. Defense 4: Ensemble of Models")

def create_ensemble_prediction(models, x):
    """Create an ensemble prediction by averaging predictions from multiple models"""
    all_predictions = []
    for model in models:
        predictions = model.predict(x)
        all_predictions.append(predictions)
    
    # Average the predictions
    ensemble_pred = np.mean(all_predictions, axis=0)
    return ensemble_pred

# Create several new models with different architectures
def create_model_1():
    model = Sequential([
        Conv2D(16, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        MaxPooling2D((2, 2)),
        Conv2D(32, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

def create_model_2():
    model = Sequential([
        Conv2D(32, (5, 5), activation='relu', input_shape=(28, 28, 1)),
        MaxPooling2D((2, 2)),
        Conv2D(64, (5, 5), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.4),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

def create_model_3():
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(28, 28, 1)),
        Conv2D(32, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Dropout(0.25),
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Dropout(0.25),
        Flatten(),
        Dense(256, activation='relu'),
        Dropout(0.5),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Create the models
model_1 = create_model_1()
model_2 = create_model_2()
model_3 = create_model_3()

# Apply data augmentation for training
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1
)

# Train model 1
print("Training ensemble model 1...")
model_1.fit(
    datagen.flow(x_train, y_train_cat, batch_size=128),
    epochs=5,
    steps_per_epoch=len(x_train) // 128,
    verbose=1
)

# Train model 2
print("Training ensemble model 2...")
model_2.fit(
    datagen.flow(x_train, y_train_cat, batch_size=128),
    epochs=5,
    steps_per_epoch=len(x_train) // 128,
    verbose=1
)

# Train model 3
print("Training ensemble model 3...")
model_3.fit(
    datagen.flow(x_train, y_train_cat, batch_size=128),
    epochs=5,
    steps_per_epoch=len(x_train) // 128,
    verbose=1
)

# Create an ensemble evaluator
class EnsembleModel:
    def __init__(self, models):
        self.models = models
    
    def predict(self, x):
        return create_ensemble_prediction(self.models, x)
    
    def evaluate(self, x, y, verbose=0):
        # Get ensemble predictions
        y_pred = self.predict(x)
        y_pred_classes = np.argmax(y_pred, axis=1)
        y_true = np.argmax(y, axis=1)
        
        # Calculate accuracy
        accuracy = np.mean(y_pred_classes == y_true)
        
        # Return dummy loss and real accuracy
        return 0.0, accuracy

# Create the ensemble model
ensemble_model = EnsembleModel([model_1, model_2, model_3])

# Save individual models
model_1.save('models/mnist_ensemble_model_1.keras')
model_2.save('models/mnist_ensemble_model_2.keras')
model_3.save('models/mnist_ensemble_model_3.keras')
print("Ensemble models saved as 'models/mnist_ensemble_model_*.keras'")

# Evaluate ensemble model
ensemble_results = evaluate_model(
    ensemble_model,
    "Ensemble Defense",
    x_test,
    y_test_cat,
    x_test_triggered,
    TARGET_LABEL
)
defense_results.append(ensemble_results)

# 6. Defense 5: Neural Cleanse (Backdoor Detection and Mitigation)
print("\n6. Defense 5: Neural Cleanse (Backdoor Detection)")

# This is a simplified version of Neural Cleanse for demonstration
# Neural Cleanse tries to find the minimum modification needed to make all inputs 
# classified as a particular target class

def find_backdoor_trigger(model, target_class, input_shape=(28, 28, 1), 
                         learning_rate=0.1, iterations=200):
    """
    Find the smallest perturbation that can trigger classification to target_class
    for a sample of images from each class
    """
    # Initialize the trigger pattern with small random values
    trigger = tf.Variable(tf.random.uniform(input_shape, 0, 0.1))
    
    # Initialize the optimization
    optimizer = tf.optimizers.Adam(learning_rate=learning_rate)
    
    # Select a small number of clean examples from each class
    samples_per_class = 3
    x_samples = []
    y_samples = []
    
    for label in range(10):
        indices = np.where(y_test == label)[0][:samples_per_class]
        x_samples.extend(x_test[indices])
        y_samples.extend([label] * len(indices))
    
    x_samples = np.array(x_samples)
    
    # Define the loss function (we want the model to classify as target_class)
    def loss_fn():
        x_backdoored = tf.clip_by_value(x_samples + trigger, 0, 1)
        predictions = model(x_backdoored)
        target_probs = predictions[:, target_class]
        classification_loss = 1.0 - tf.reduce_mean(target_probs)
        
        # Add L1 regularization to encourage sparsity
        l1_loss = 0.1 * tf.reduce_sum(tf.abs(trigger))
        
        return classification_loss + l1_loss
    
    # Optimization loop
    loss_history = []
    for i in range(iterations):
        with tf.GradientTape() as tape:
            loss = loss_fn()
        
        gradients = tape.gradient(loss, [trigger])
        optimizer.apply_gradients(zip(gradients, [trigger]))
        
        # Project to valid image space
        trigger.assign(tf.clip_by_value(trigger, 0, 1))
        
        loss_history.append(loss.numpy())
        
        if i % 20 == 0:
            print(f"Iteration {i}, Loss: {loss.numpy():.4f}")
    
    # Return the final trigger pattern
    return trigger.numpy()

# Find potential triggers for each class
print("Searching for potential backdoor triggers...")
potential_triggers = []
trigger_losses = []

for target_class in range(10):
    print(f"Finding trigger for target class {target_class}...")
    trigger = find_backdoor_trigger(backdoored_model, target_class, iterations=100)
    
    # Apply the trigger to test samples and check success rate
    x_test_with_trigger = np.clip(x_test + trigger, 0, 1)
    preds = backdoored_model.predict(x_test_with_trigger)
    pred_classes = np.argmax(preds, axis=1)
    success_rate = np.mean(pred_classes == target_class)
    
    # Calculate L1 norm of the trigger (measure of its size)
    l1_norm = np.sum(np.abs(trigger))
    
    print(f"Class {target_class}: Trigger success rate: {success_rate:.4f}, L1 norm: {l1_norm:.4f}")
    
    potential_triggers.append(trigger)
    trigger_losses.append(l1_norm / success_rate)  # Anomaly index

# Identify the potential backdoored class
outlier_index = np.argmin(trigger_losses)
print(f"Potential backdoored class: {outlier_index} (Anomaly index: {trigger_losses[outlier_index]:.4f})")

# Visualize all potential triggers
plt.figure(figsize=(15, 8))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(potential_triggers[i].reshape(28, 28), cmap='viridis')
    plt.title(f"Class {i}: {trigger_losses[i]:.2f}")
    plt.axis('off')

plt.tight_layout()
plt.savefig('defense_results/neural_cleanse_triggers.png')
print("Neural Cleanse potential triggers saved to 'defense_results/neural_cleanse_triggers.png'")

# 7. Compare all defense methods
print("\n7. Comparing all defense methods...")

# Create a comparison table
defense_names = [r['name'] for r in defense_results]
clean_accuracies = [r['clean_accuracy'] for r in defense_results]
backdoor_rates = [r['backdoor_success_rate'] for r in defense_results]

# Plot comparison
plt.figure(figsize=(12, 6))
x = np.arange(len(defense_names))
width = 0.35

plt.bar(x - width/2, [acc * 100 for acc in clean_accuracies], width, 
        label='Clean Test Accuracy (%)')
plt.bar(x + width/2, [rate * 100 for rate in backdoor_rates], width,
        label='Backdoor Success Rate (%)')

plt.xlabel('Defense Method')
plt.ylabel('Percentage (%)')
plt.title('Comparison of Backdoor Defense Methods')
plt.xticks(x, [name[:15] + '...' if len(name) > 15 else name for name in defense_names], rotation=45, ha='right')
plt.legend()

# Add value labels
for i, v in enumerate(clean_accuracies):
    plt.text(i - width/2, v * 100 + 1, f"{v*100:.1f}%", ha='center')

for i, v in enumerate(backdoor_rates):
    plt.text(i + width/2, v * 100 + 1, f"{v*100:.1f}%", ha='center')

plt.tight_layout()
plt.savefig('defense_results/defense_comparison.png')
print("Defense method comparison saved to 'defense_results/defense_comparison.png'")

# 8. Generate defense report
print("\n8. Generating defense report...")

report = f"""
BACKDOOR DEFENSE TECHNIQUES REPORT
==================================

1. DEFENSE SUMMARY
-----------------
This report compares different defense techniques against a backdoor attack on an MNIST digit classifier.
The backdoor was trained to classify images with a small {TRIGGER_SIZE}x{TRIGGER_SIZE} pixel pattern at position {TRIGGER_LOCATION} as digit {TARGET_LABEL}.

2. DEFENSE METHODS EVALUATED
--------------------------
1. Input Preprocessing (Noise + Blur): Adding noise and blurring to disrupt the trigger pattern
2. Adversarial Training: Training with a mix of clean and triggered examples (keeping original labels)
3. Fine-Pruning: Pruning {prune_percentage*100:.1f}% of the weights and fine-tuning on clean data
4. Ensemble of Models: Using majority voting from multiple independently trained models
5. Neural Cleanse: Detecting and identifying the potential backdoor

3. RESULTS COMPARISON
-------------------
Defense Method                  | Clean Accuracy | Backdoor Success Rate | Effectiveness
-------------------------------|----------------|----------------------|---------------
"""

# Add each defense to the report
for result in defense_results:
    # Calculate effectiveness as reduction in backdoor success rate with minimal clean accuracy loss
    if result['name'] == "Backdoored Model (No Defense)":
        baseline_accuracy = result['clean_accuracy']
        baseline_backdoor_rate = result['backdoor_success_rate']
        effectiveness = "Baseline"
    else:
        accuracy_drop = baseline_accuracy - result['clean_accuracy']
        backdoor_reduction = baseline_backdoor_rate - result['backdoor_success_rate']
        
        if backdoor_reduction > 0.5 and accuracy_drop < 0.05:
            effectiveness = "High"
        elif backdoor_reduction > 0.2 and accuracy_drop < 0.1:
            effectiveness = "Medium"
        else:
            effectiveness = "Low"
    
    report += f"{result['name'][:28]:<30} | {result['clean_accuracy']*100:6.2f}%      | {result['backdoor_success_rate']*100:6.2f}%              | {effectiveness}\n"

report += f"""
4. KEY FINDINGS
-------------
- {'The most effective defense in terms of reducing backdoor success rate was ' + min(defense_results[1:], key=lambda x: x['backdoor_success_rate'])['name']}
- {'The defense with the least impact on clean accuracy was ' + max(defense_results[1:], key=lambda x: x['clean_accuracy'])['name']}
- {'Neural Cleanse correctly identified class ' + str(outlier_index) + ' as the potential backdoored class' if outlier_index == TARGET_LABEL else 'Neural Cleanse identified class ' + str(outlier_index) + ' as the potential backdoored class (actual: ' + str(TARGET_LABEL) + ')'}

5. RECOMMENDATIONS
---------------
1. For deployment:
   - Use a combination of preprocessing and ensemble models for the most robust defense
   - Implement input sanitization to detect and remove potential triggers
   - Regularly retrain models on verified clean data

2. For development:
   - Use adversarial training during model development
   - Implement fine-pruning before deployment
   - Apply Neural Cleanse or similar techniques to detect potential backdoors

3. For ongoing monitoring:
   - Regularly scan for anomalies in model behavior
   - Test with specially crafted inputs to detect possible backdoors
   - Monitor distribution shifts in predictions
"""

with open('defense_results/defense_techniques_report.txt', 'w') as f:
    f.write(report)

print("Defense techniques report saved to 'defense_results/defense_techniques_report.txt'")

# 9. Create a comprehensive dashboard for defense results
print("\n9. Creating defense results dashboard...")

plt.figure(figsize=(15, 10))

# Title and summary
plt.subplot(3, 3, 1)
plt.axis('off')
plt.text(0.5, 0.9, 'MNIST Backdoor Defense Results', ha='center', fontsize=18, fontweight='bold')
plt.text(0.5, 0.7, f'Original Backdoor Target: Class {TARGET_LABEL}', ha='center', fontsize=14)
plt.text(0.5, 0.6, f'Original Backdoor Success Rate: {backdoor_success_rate*100:.1f}%', ha='center', fontsize=14)
plt.text(0.5, 0.5, f'Best Defense: {min(defense_results[1:], key=lambda x: x["backdoor_success_rate"])["name"]}', ha='center', fontsize=14)
plt.text(0.5, 0.4, f'Highest Clean Accuracy: {max(defense_results[1:], key=lambda x: x["clean_accuracy"])["name"]}', ha='center', fontsize=14)

# Defense comparison - Clean accuracy vs Backdoor success rate
plt.subplot(3, 3, 2)
for result in defense_results:
    if result['name'] == "Backdoored Model (No Defense)":
        color = 'red'
        marker = 'X'
        size = 100
    else:
        color = 'blue'
        marker = 'o'
        size = 80
    
    plt.scatter(result['clean_accuracy']*100, result['backdoor_success_rate']*100, 
               color=color, s=size, alpha=0.7, marker=marker)
    plt.text(result['clean_accuracy']*100 + 0.5, result['backdoor_success_rate']*100 + 1, 
             result['name'][:10], fontsize=8)

plt.title('Defense Methods Comparison')
plt.xlabel('Clean Test Accuracy (%)')
plt.ylabel('Backdoor Success Rate (%)')
plt.xlim(90, 100)
plt.ylim(0, 100)
plt.grid(True, linestyle='--', alpha=0.7)

# Show trigger pattern detected by Neural Cleanse
plt.subplot(3, 3, 3)
plt.imshow(potential_triggers[outlier_index].reshape(28, 28), cmap='viridis')
plt.title(f'Detected Backdoor Trigger (Class {outlier_index})')
plt.axis('off')

# Show effect of preprocessing
plt.subplot(3, 3, 4)
plt.axis('off')
plt.title('Effect of Input Preprocessing')

# Display a few examples of preprocessing defense
for i in range(4):
    # Original triggered image
    plt.subplot(3, 8, 9 + i)
    plt.imshow(x_test_triggered[i].reshape(28, 28), cmap='gray')
    plt.title(f"Original", fontsize=10)
    plt.axis('off')
    
    # Preprocessed image
    preprocessed_images = apply_input_preprocessing(
        x_test_triggered[i:i+1], 
        noise_std=0.1, 
        blur_kernel_size=3
    )
    plt.subplot(3, 8, 13 + i)
    plt.imshow(preprocessed_images[0].reshape(28, 28), cmap='gray')
    plt.title("Preprocessed", fontsize=10)
    plt.axis('off')

# Bar chart of defense effectiveness
plt.subplot(3, 3, 5)
# Calculate effectiveness scores (normalized reduction in backdoor success rate)
baseline_backdoor_rate = defense_results[0]['backdoor_success_rate']
effectiveness_scores = []

for result in defense_results[1:]:  # Skip the backdoored model
    # Calculate how much the backdoor success rate was reduced
    reduction = baseline_backdoor_rate - result['backdoor_success_rate']
    # Normalize by the original rate
    normalized_reduction = reduction / baseline_backdoor_rate
    effectiveness_scores.append(normalized_reduction * 100)  # As percentage

defense_names = [r['name'].replace('Defense', '') for r in defense_results[1:]]
plt.bar(range(len(effectiveness_scores)), effectiveness_scores)
plt.title('Defense Effectiveness')
plt.ylabel('Backdoor Reduction (%)')
plt.xticks(range(len(effectiveness_scores)), [name[:10] for name in defense_names], rotation=45, ha='right')
plt.ylim(0, 100)

# Show defense method strengths and weaknesses
plt.subplot(3, 3, 6)
plt.axis('off')
plt.title('Defense Methods Overview')

overview_text = """
Input Preprocessing:
+ Easy to implement
+ No retraining required
- May affect clean accuracy

Adversarial Training:
+ Robust against known triggers
+ Good clean accuracy
- Requires knowledge of attack

Fine-Pruning:
+ Removes backdoor connections
+ Preserves clean accuracy
- Requires retraining

Ensemble:
+ Robust solution
+ Good overall performance
- Computationally expensive

Neural Cleanse:
+ Detects unknown backdoors
+ Identifies target class
- Complex implementation
"""
plt.text(0, 0.1, overview_text, fontsize=9, va='top', ha='left')

# Training curves comparison
plt.subplot(3, 3, 7)
plt.title('Training Accuracy Curves')
plt.plot(adv_history.history['accuracy'], label='Adv. Training')
plt.plot(fine_tune_history.history['accuracy'], label='Fine-Pruning')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Show misclassified examples
plt.subplot(3, 3, 8)
plt.axis('off')
plt.title('Ensemble Predictions on Triggered Images')

# Find examples where ensemble defense works
ensemble_preds = ensemble_model.predict(x_test_triggered)
ensemble_pred_classes = np.argmax(ensemble_preds, axis=1)
backdoored_pred_classes = np.argmax(backdoored_model.predict(x_test_triggered), axis=1)

# Find examples where ensemble works but backdoored model fails
success_indices = []
for i in range(len(x_test_triggered)):
    if ensemble_pred_classes[i] != TARGET_LABEL and backdoored_pred_classes[i] == TARGET_LABEL:
        success_indices.append(i)
        if len(success_indices) >= 4:
            break

for i, idx in enumerate(success_indices[:4]):
    plt.subplot(3, 8, 17 + i)
    plt.imshow(x_test_triggered[idx].reshape(28, 28), cmap='gray')
    plt.title(f"True: {y_test[idx]}", fontsize=10)
    plt.axis('off')
    
    plt.subplot(3, 8, 21 + i)
    plt.text(0.5, 0.7, f"Backdoored: {backdoored_pred_classes[idx]}", fontsize=10, ha='center')
    plt.text(0.5, 0.3, f"Ensemble: {ensemble_pred_classes[idx]}", fontsize=10, ha='center', color='green')
    plt.axis('off')

# Final recommendations
plt.subplot(3, 3, 9)
plt.axis('off')
plt.title('Recommended Defense Strategy')
recommendations = """
1. Data Validation:
   • Verify training data sources
   • Use anomaly detection

2. Model Protection:
   • Implement preprocessing
   • Use model ensembles
   • Apply fine-pruning

3. Deployment Security:
   • Regular backdoor scanning
   • Monitor prediction distributions
   • Test with potential triggers
"""
plt.text(0, 0.1, recommendations, fontsize=9, va='top', ha='left')

plt.tight_layout()
plt.savefig('defense_results/defense_dashboard.png', dpi=300)
print("Defense dashboard saved to 'defense_results/defense_dashboard.png'")

print("\n======== BACKDOOR DEFENSE EVALUATION COMPLETED ========")
print("Defense methods evaluated:")
for i, result in enumerate(defense_results):
    print(f"{i+1}. {result['name']}")
    print(f"   Clean Accuracy: {result['clean_accuracy']*100:.2f}%")
    print(f"   Backdoor Success Rate: {result['backdoor_success_rate']*100:.2f}%")
print("All defense results saved to 'defense_results/'")
print("========================================================")
```
